```python
import re
import pandas as pd
import io

# PASTE YOUR TEXT FILE CONTENT BETWEEN THE TRIPLE QUOTES BELOW
raw_text = """
3/4/2022 11:54 AM
@McKinney Bay Knights (Ashtin) :MCK: acquire rights to Bo Bichette from @Fort Worth Fire (Austin) :FW: in exchange for rookie contract eligible Tarik Skubal, rights to Cedrick Mullins, prospect Hudson Head, an $123 of 2022 Griffey Coin
3/4/2022 4:41 PM
@Amherst Minutemen :AMH: acquire rights to Matt Brash from @Fort Worth Fire (Austin) :FW: in exchange for $25.00 of 2022 Griffey Coin.
3/8/2022 10:31 AM
@Chicago Communists (Keith) :CHI: acquire prospect Garrett Mitchell from @Amherst Minutemen :AMH: in exchange for $5.00 of 2022 Griffey Coin.
3/8/2022 4:53 PM
@Beamsville Beavers (Jacob) :SC: acquire prospect Drew Waters from @Amherst Minutemen :AMH: in excahnge for $1.00 of 2023 Griffey Coin.
@McKinney Bay Knights (Ashtin) :MCK: acquire Rhys Hoskins, rights to Giancarlo Stanton, and rights to Brandon Crawford from @Brooklyn Brewers (Nic) :BRK: in exchange for the a 2022 1st Rd OP (MCK, 4th OVR), a 2022 1st Rd OP (BRK, 10th OVR), and $10.00 of 2023 Griffey Coin. 
3/11/2022 7:16 PM
@Amherst Minutemen :AMH: acquire Kenta Maeda and  $17 of 2023 Griffey Coin from @Fort Worth Fire (Austin) :FW:. 
3/15/2022 8:53 AM
@Halton Huskies (Brendan) :HAL: acquire rights to Kevin Gausman from @Andre - LA Smog :LA: in exchange for $2.00 of 2022 Griffey Coin. 
@Andre - LA Smog :LA: acquire prospects Luis Gil and Seth Beer from @Wachusett Mountaineers :COL: in exchange for $2.00 of 2022 Griffey Coin.
3/17/2022 3:57 PM
@Amherst Minutemen :AMH: acquire Carlos Martinez, Yoan Mocada, and prospect George Valera from @Toronto Canucks :TOR: (no return).
@Andre - LA Smog :LA: acquire prospect George Valera from @Amherst Minutemen :AMH:  in exchange for rights to Bryon Buxton and $10.00 of 2023 Griffey Coin.
3/18/2022 9:20 AM
@Amherst Minutemen :AMH: acquire Mookie Betts and prospect Curtis Mead from @Andre - LA Smog :LA: in exchange for prospect Vidal Brujan and $15.00 of 2023 Griffey Coin.
@Beamsville Beavers (Jacob) :SC: acquire Mookie Betts (50% retained) and prospect Hunter Greene from @Amherst Minutemen :AMH: in exchange for prospects Grayson Rodriguez and rights to Eloy Jimenez.
3/18/2022 9:55 AM
@McKinney Bay Knights (Ashtin) :MCK: acquire rights to Hunter Renfroe, rights to Myles Straw, rights to Jeimer Candelario, rights to Amed Rosario, and rights to Mike Zunino from @Amherst Minutemen :AMH: in exchange for $15.00 of 2023 Griffey Coin.
3/18/2022 10:03 AM
@McKinney Bay Knights (Ashtin) :MCK: acquire rights to Kyle Seager from @Chicago Communists (Keith) :CHI: in exchange for $1.00 of 2023 Griffey Coin.
3/18/2022 10:13 AM
@McKinney Bay Knights (Ashtin) :MCK: acquire rights to Luis Urias and rights to Yohel Pozo from @Wachusett Mountaineers :COL: in exchange for $4.00 of 2023 Griffey Coin.
3/18/2022 2:21 PM
@Zach :NO: acquire rights to Eloy Jimenez from @Amherst Minutemen :AMH: in exchange for $35.00 of 2023 Griffey Coin.
@Wachusett Mountaineers :COL: acquire Corey Kluber from @Zach :NO: in exchange for $1.00 of 2023 Griffey Coin. 
3/18/2022 3:04 PM
@McKinney Bay Knights (Ashtin) :MCK: acquire rights to Omar Navarez from @Toronto Canucks :TOR: in exchange for $1.00 of 2023 Griffey Coin.
3/21/2022 8:15 AM
@Oklahoma Schooners (Matt) :OK: acquire rights to Logan Webb from @Andre - LA Smog :LA: in exchange for $9.00 of 2023 Griffey Coin.
3/21/2022 10:03 AM
@Amherst Minutemen :AMH: acquire Tyler Glasnow and $20.00 of 2023 Griffey Coin from @Halton Huskies (Brendan) :HAL: in exchange for the rights to James Paxton.
3/22/2022 9:15 AM
@Amherst Minutemen :AMH: acquire rights to Nicky Lopez and rights to Adam Frazier from @Chicago Communists (Keith) :CHI: in exchange for $4.00 of 2023 Griffey Coin.
3/23/2022 8:29 AM
@Fort Worth Fire (Austin) :FW: acquire Gavin Lux, Dylan Carlson, Eduardo Rodriguez, rights to Ty France, and rights to Marcus Semien from @Beamsville Beavers (Jacob) :SC: in exchange for Manny Machado (50% retained), Jameson Taillon (50% retained), rights to DJ LeMahieu, and $8.00 of 2023 Griffey Coin.
3/23/2022 2:28 PM
@McKinney Bay Knights (Ashtin) :MCK: acquire rights to Fernando Tatis Jr. and rights to Tommy Edman from @typo21 :DAL: in exchange for Rhys Hoskins, Tanner Houck, rights to Hunter Renfroe, rights to Brandon Crawford and $19.00 of 2023 Griffey Coin.
3/23/2022 2:36 PM
@Oklahoma Schooners (Matt) :OK: acquire rights to Anthony DeSclafani from @typo21 :DAL: in exchange for a 2022 3rd Rd OP (OK, 33rd Overall) and $2.00 of 2023 Griffey Coin.
3/24/2022 10:52 AM
@Oklahoma Schooners (Matt) :OK: acquire rights to Carlos Rodon and rights to Alex Wood from @Brooklyn Brewers (Nic) :BRK: in exchange for rights to Kyle Gibson and rights to Wade Miley.
3/24/2022 11:20 AM
@Amherst Minutemen :AMH: acquire rights to Ty France from @Fort Worth Fire (Austin) :FW: in exchange for rights to Mike Clevinger.
3/24/2022 11:27 AM
@Wachusett Mountaineers :COL: acquire Kyle Lewis and $35.00 of 2023 Griffey Coin from @Amherst Minutemen :AMH: (no return).
3/31/2022 3:06 PM
@Fort Worth Fire (Austin) :FW: acquire a 2022 3rd Round OP (BRK, 38th OVR) and a 2022 4th Round OP (AMH, 43rd OVR) from @Amherst Minutemen :AMH: in exchange for $1.00 of 2023 Griffey Coin.
@Halton Huskies (Brendan) :HAL: acquire a 2022 3rd Round OP (SC, 40th OVR) and a 2022 4th Round OP (SC, 54th OVR) from @Beamsville Beavers (Jacob) :SC: in exchange for $1.00 of 2023 Griffey Coin.
4/1/2022 9:16 AM
@Zach :NO: acquire Zack Wheeler (50% retained), Yu Darvish (50% retention), and Christian Hernandez from @Wachusett Mountaineers :COL: in exchange for Andrew Benentendi, Nick Senzel, Daniel Lynch prospect Brady House, prospect Colton Cowser, $16 of 2023 Griffey Coin, and 2023 1st Rd OP (NO).
4/1/2022 1:15 PM
@Toronto Canucks :TOR: acquire Eddie Rosario from @Wachusett Mountaineers :COL: in exchange for prospect Greg Jones.
@Amherst Minutemen :AMH: acquire Brandon Lowe and Carlos Correa from @Wachusett Mountaineers :COL: in exchange for Spencer Howard, prospect Nick Gonzales, $40.00 of 2023 Griffey Coin, and a 2023 2nd Rd OP (AMH).
4/3/2022 12:22 PM
@Andre - LA Smog :LA: acquire a 2022 4th Rd OP (FW, 56th OVR) from @Fort Worth Fire (Austin) :FW: in exchange for $1.00 of 2023 Griffey Coin. 
@Toronto Canucks :TOR: acquire a 2022 3rd Rd OP (AMH, 29th OVR) from @Amherst Minutemen :AMH: in exchange for $1.00 of 2023 Griffey Coin.
4/3/2022 12:30 PM
@Andre - LA Smog :LA: acquire Brady Singer from @Chicago Communists (Keith) :CHI: in exchange for $2.00 of 2023 Griffey Coin.
4/5/2022 2:21 PM
@Oklahoma Schooners (Matt) :OK: acquire Jake Odorizzi from @Fort Worth Fire (Austin) :FW: in exchange for Andrew McCutchen and $6.00 of 2023 Griffey Coin. 
@Fort Worth Fire (Austin) :FW: acquire Sean Manaea from @McKinney Bay Knights (Ashtin) :MCK: in exchange for prospect Daulton Jefferies and $5.00 of 2023 Griffey Coin. 
@Zach :NO: acquire Nestor Corets and Tyler Stephenson from @Amherst Minutemen :AMH: in exchange for Patrick Sandoval.
@Beamsville Beavers (Jacob) :SC: acquire Eric Lauer from @Amherst Minutemen :AMH: in exchange for Elias Diaz and $4.00 of 2023 Griffey Coin.
@Zach :NO: acquire Carter Kieboom from @Amherst Minutemen :AMH: in exchange for $1.00 of 2023 Griffey Coin.
4/6/2022 8:50 AM
@Amherst Minutemen :AMH: acquire Chris Flexen from @McKinney Bay Knights (Ashtin) :MCK: in exchange for Nico Hoerner and $3.00 of 2023 Griffey Coin.
4/6/2022 4:59 PM
@typo21 :DAL: acquire Nathan Eovaldi from @Andre - LA Smog :LA: in exchange for prospect Nick Yorke.
4/6/2022 7:07 PM
@Brooklyn Brewers (Nic) :BRK: acquire Dakota Hudson from @Wachusett Mountaineers :COL: in exchange for a 2023 3rd Rd OP (BRK).
4/6/2022 7:20 PM
@Amherst Minutemen :AMH: acquire Logan Webb (50% retained) from @Wachusett Mountaineers :COL: in exchange for prospect Zac Veen.
4/6/2022 10:04 PM
@McKinney Bay Knights (Ashtin) :MCK: acquire Kyle Freeland from @Wachusett Mountaineers :COL: in exchange for Anibal Sanchez and $1.00 of 2023 Griffey Coin.
4/8/2022 12:34 PM
@Fort Worth Fire (Austin) :FW: acquire J.D. Martinez and Giancarlo Stanton from @McKinney Bay Knights (Ashtin) :MCK: in exchange for Dylan Carlson, Andrew McCutchen, and prospect Hudson Head.
4/11/2022 11:19 AM
@McKinney Bay Knights (Ashtin) :MCK: acquire Chris Paddack from @Wachusett Mountaineers :COL: in exchange for Kyle Freeland.
4/12/2022 2:24 PM
@McKinney Bay Knights (Ashtin) :MCK: acquire Tommy Romero from @Andre - LA Smog :LA: in exchange for $3.00 of 2023 Griffey Coin.
4/17/2022 6:24 PM
@Amherst Minutemen :AMH: acquire Kyle Tucker from @Fort Worth Fire (Austin) :FW: in exchange for Josiah Gray, Alex Kirilloff,  prospect Anthony Vople, and $30.00 of 2022 Griffey Coin.
4/20/2022 3:03 PM
@Wachusett Mountaineers :COL: acquire Chris Archer and Rich Hill from @McKinney Bay Knights (Ashtin) :MCK: in exchange for Corey Kluber.
4/20/2022 3:40 PM
@McKinney Bay Knights (Ashtin) :MCK: acquire Drew Rasmussen from @Andre - LA Smog :LA: in exchange for $4.00 of 2023 Griffey Coin.
4/21/2022 8:43 AM
@Toronto Canucks :TOR: acquire Michael Wacha from @McKinney Bay Knights (Ashtin) :MCK: in exchange for Josh Flemming and $4.00 of 2023 Griffey Coin.
@Halton Huskies (Brendan) :HAL: acquire Cal Quantrill from @McKinney Bay Knights (Ashtin) :MCK: in exchange for Ryan Yarbrough.
4/25/2022 9:36 AM
@Wachusett Mountaineers :COL: acquire Bailey Ober from the @Andre - LA Smog :LA: in exchange for $4.00 of 2023 Griffey Coin.
@Amherst Minutemen :AMH: acquire J.P. Crawford from @Wachusett Mountaineers :COL: in exchange for $5.00 of 2023 Griffey Coin.
@Amherst Minutemen :AMH: acquire Austin Nola from @Fort Worth Fire (Austin) :FW: in excahnge for Elias Diaz and $5.00 of 2023 Griffey Coin.
4/25/2022 11:23 AM
@Brooklyn Brewers (Nic) :BRK: acquire Salvador Perez, Luis Urias, Wade Miley, Mitch Keller, prospect Edward Cabrera, and prospect JJ Bleday from @McKinney Bay Knights (Ashtin) :MCK: in exchange for Shane McClanahan, Gary Sanchez, and Patrick Wisdom.
4/25/2022 2:05 PM
@Amherst Minutemen :AMH: acquire Jurickson Profar from @McKinney Bay Knights (Ashtin) :MCK: in exchange for Tyler Glasnow and prospect Curtis Mead.
4/25/2022 3:55 PM
@Toronto Canucks :TOR: acquire Justin Turner from @McKinney Bay Knights (Ashtin) :MCK: in exchange for prospect Jose Miranda.
4/26/2022 8:41 AM
@Toronto Canucks :TOR: acquire Jordan Lyles from @Wachusett Mountaineers :COL: in exchange for $4.00 of 2023 Griffey Coin.
4/26/2022 1:51 PM
@Beamsville Beavers (Jacob) :SC: acquire Lourdes Gurriel Jr. from @Wachusett Mountaineers:COL:  in exchange for prospect Eddys Lenard and $4.00 of 2023 Griffey Coin.
4/26/2022 2:05 PM
@Chicago Communists (Keith) :CHI: acquire Chad Kuhl from @Wachusett Mountaineers :COL: in exchange for $5.00 of 2023 Griffey Coin.
4/26/2022 3:39 PM
@McKinney Bay Knights (Ashtin) :MCK: acquire Tom Murphy from @Wachusett Mountaineers :COL: in exchange for $2.00 of 2023 Griffey Coin.
5/2/2022 1:01 PM
@Fort Worth Fire (Austin) :FW: acquire Andrew Benintendi from @Wachusett Mountaineers :COL: in exchange for Trent Grisham, prospect Justin Foscue, and $15.00 of 2023 Griffey Coin.
5/5/2022 11:37 AM
@Oklahoma Schooners (Matt) :OK: acquire Chris Archer from @Wachusett Mountaineers :COL: in exchange for $2.00 of 2023 Griffey Coin.
5/9/2022 12:20 PM
@Andre - LA Smog :LA: acquire Chris Paddack (DFA team) from @McKinney Bay Knights (Ashtin) :MCK: in exchange for Josh Rojas.
5/12/2022 8:22 AM
@Chicago Communists (Keith) :CHI: acquire Trae Turner from @Princeton Anglers (GM - Dain) @Sanger Vientos - Brandon :PRI: in exchange for prospect O'neil Cruz, prospect Jackson Jobe, prospect Nick Pratto, and $20.00 of 2023 Griffey Coin.
5/13/2022 10:26 AM
@McKinney Bay Knights (Ashtin) :MCK: acquire Owen Miller from @Wachusett Mountaineers :COL: in exchange for Adolis Garcia and prospect Hudson Head.
5/13/2022 12:17 PM
@Andre - LA Smog :LA: acquire Jared Walsh (DFA) from @Amherst Minutemen :AMH: in exchange for $5.00 of 2023 Griffey Coin.
5/13/2022 12:38 PM
@Beamsville Beavers (Jacob) :SC: acquire Bryce Harper and $5.00 of 2023 Griffey Coin from @typo21 :DAL: in exchange for prospect Jasson Dominguez and prospect Hunter Greene.
5/13/2022 2:13 PM
@Beamsville Beavers (Jacob) :SC: acquire Jean Segura from @Toronto Canucks :TOR: in exchange for prospect Pedro Leon and $3.00 of 2023 Griffey Coin.
5/16/2022 8:13 AM
@Amherst Minutemen :AMH: acquire Christian Yelich from @Toronto Canucks :TOR: in exchange for prospect Gabriel Montero and $5.00 of 2023 Griffey Coin.
@Halton Huskies (Brendan) :HAL: acquire Jose Berrios, Hyun Jin Ryu, and Yusei Kikuchi and $10.00 of 2023 Griffey Coin from @Toronto Canucks :TOR: in exchange for prospect Max Meyer and prospect Everson Pereira. 
5/16/2022 9:48 AM
@Andre - LA Smog :LA: acquire Nathan Eovaldi from @typo21 :DAL: in exchange for prospect Vidal Brujan and $6.00 of 2023 Griffey Coin.
5/16/2022 11:10 AM
@Amherst Minutemen :AMH: acquire Noah Syndergaard from @Princeton Anglers (GM - Dain) @Sanger Vientos - Brandon :PRI: in exchange for Matt Manning and $10.00 of 2023 Griffey Coin.
5/17/2022 8:07 PM
@Halton Huskies (Brendan) :HAL: acquire Max Scherzer ($4.00 retained) and Yordan Alvarez ($5.00 retained from @typo21 :DAL: in exchange for Tony Goslin, Nick Pivetta, prospect Henry Davis, prospect Liover Pegeuro, prospect Joey Wiemer, prospect Jeter Downs, prospect Jordan Balazovic, and $15.00 of 2023 Griffey Coin.
5/18/2022 9:16 AM
@Toronto Canucks :TOR: acquire Dylan Bundy from @Halton Huskies (Brendan) :HAL: in exchange for $3.00 of 2023 Griffey Coin.
5/19/2022 10:18 AM
@McKinney Bay Knights (Ashtin) :MCK: acquire Juan Soto (50% retained, $9.25) from @Princeton Anglers (GM - Dain) @Sanger Vientos - Brandon :PRI: in exchange for prospect Triston Casas, prospect Corbin Carroll, prospect Nolan Gorman, and prospect Nate Pearson.
5/20/2022 1:54 PM
@Fort Worth Fire (Austin) :FW: acquire Tony Gonsolin and $5.00 of 2023 Griffey coin from @typo21 :DAL:  in exchange for prospect Ty Madden and prospect Jarren Duran. 
5/23/2022 8:27 AM
@Brooklyn Brewers (Nic) :BRK: acquire Michael Lorenen from @Andre - LA Smog :LA: in exchange for $7.00 of 2023 Griffey Coin.
@Brooklyn Brewers (Nic) :BRK: acquire Jakob Junis from @Andre - LA Smog :LA: in exchange for $5.00 of 2023 Griffey Coin.
@McKinney Bay Knights (Ashtin) :MCK: acquire Drew Smyly from @Oklahoma Schooners (Matt) :OK: in exchange for $1.00 of 2023 Griffey Coin.
5/26/2022 3:18 PM
@Brooklyn Brewers (Nic) :BRK: acquire Konnor Pilkington from @McKinney Bay Knights (Ashtin) :MCK: in exchange for $1.00 of 2023 Griffey Coin.
5/31/2022 10:17 AM
@Halton Huskies (Brendan) :HAL: acquire Jeff McNeil from @Toronto Canucks :TOR: in exchange for a 2023 2nd Rd OP (HAL) and $10.00 of 2023 Griffey Coin. 
6/6/2022 12:38 PM
@Oklahoma Schooners (Matt) :OK: acquire Logan Webb ($1.50 retained) from @Amherst Minutemen :AMH: in exchange for prospect Spencer Torkelson and prospect Kyler Murray.
6/6/2022 1:23 PM
@Fort Worth Fire (Austin) :FW: acquire Gary Sanchez from @McKinney Bay Knights (Ashtin) :MCK: in exchange for a 2023 2nd Rd OP (FW).
6/6/2022 1:38 PM
@McKinney Bay Knights (Ashtin) :MCK: acquire Jared Walsh from @Andre - LA Smog :LA: in exchange for Abraham Toro.
6/8/2022 9:10 AM
@McKinney Bay Knights (Ashtin) :MCK: acquire Mike Minor from @Oklahoma Schooners (Matt) :OK: in exchange for $1.00 of 2023 Griffey Coin.
@Beamsville Beavers (Jacob) :SC: acquire Adrian Houser from @Fort Worth Fire (Austin) :FW: in exchange for a 2023 2nd Rd OP (SC) and $3.00 of 2023 Griffey Coin.
6/9/2022 8:15 AM
@Chicago Communists (Keith) :CHI: acquire Clayton Kershaw from @Andre - LA Smog :LA: in exchange for prospect Andy Pages and $12.00 of 2023 Griffey Coin.
6/10/2022 9:31 PM
@Fort Worth Fire (Austin) :FW: acquire Francisco Lindor, George Springer, and Andrew Heaney from @Andre - LA Smog :LA: in exchange for prospect Marcelo Mayer, prospect Deigo Cartaya, prospect Sam Bachman, Alex Kirilloff, Willy Adames, Chris Taylor a 2023 1st Rd OP (FW) and $50 of 2023 Griffey Coin.
6/13/2022 6:07 PM
@Fort Worth Fire (Austin) :FW: acquire Joe Musgrove from @Andre - LA Smog :LA: in exchange for prospect Josh Jung.
6/15/2022 8:25 AM
@Andre - LA Smog :LA: acquire Nick Pivetta from @typo21 :DAL: in exchange for $8.00 of 2023 Griffey Coin.
7/19/2022 9:14 AM
@Andre - LA Smog :LA: acquire Iann Happ from @Wachusett Mountaineers :COL: in exchange for $3.00 of 2023 Griffey Coin.
@Zach :NO: acquire Christian Yelich from @Amherst Minutemen :AMH: in exchange for prospect Cristhian Vaquero.
@Beamsville Beavers (Jacob) :SC: acquire Jurickson Profar from @Amherst Minutemen :AMH: in exchange for prospect Zack Gelof.
@Beamsville Beavers (Jacob) :SC: acquire Nelson Cruz, Cristian Javier, and German Marquez from @Princeton Anglers (GM - Dain) @Sanger Vientos - Brandon :PRI: in exchange for Glenn Otto, Tommy Pham, Ryan Mountcastle and $15.00 of 2023 Griffey Coin.
7/19/2022 9:21 AM
@Beamsville Beavers (Jacob) :SC: acquire Noah Syndergaard from @Amherst Minutemen :AMH: in exchange for prospect Kody Hoese and $3.00 of 2023 Griffey Coin.
@Brooklyn Brewers (Nic) :BRK: acquire Shane Bieber ($6.00 retained), Aaron Civale ($3.25 retained), Zach Plesac ($4.25 retained) from @Andre - LA Smog :LA: in exchange for James Paxton, prospect Eury Perez, prospect JJ Bleday, prospect Ryan Pepiot, and $10.00 of 2023 Griffey Coin.
@Brooklyn Brewers (Nic) :BRK: acquire Carlos Correa from @Amherst Minutemen :AMH: in exchange for prospect Dustin Harris and $10.00 of 2023 Griffey Coin.
@Brooklyn Brewers (Nic) :BRK: acquire Josh Bell and Julio Urias from @Princeton Anglers (GM - Dain) @Sanger Vientos - Brandon :PRI: in exchange for Oscar Gonzalez, prospect Sal Frelick, prospect David Hamilton, and $22.00 of 2023 Griffey Coin.
@Halton Huskies (Brendan) :HAL: acquire Ross Stripling from @Andre - LA Smog :LA: in exchange for $4.00 of 2023 Griffey Coin.
7/19/2022 9:28 AM
@Andre - LA Smog :LA: acquire Adolis Garcia from @Wachusett Mountaineers :COL: in exchange for $3.00 of 2023 Griffey Coin.
@Fort Worth Fire (Austin) :FW: acquire Will Smith and $3.00 of 2023 Griffey Coin from @typo21 :DAL: in exchange for Josiah Gray, Casey Mize, Gary Sanchez, and a 2023 2nd Rd OP (SC). 
@Beamsville Beavers (Jacob) :SC: acquire Charlie Blackmon from @Princeton Anglers (GM - Dain) @Sanger Vientos - Brandon :PRI: in exchange for $7.00 of 2023 Griffey Coin.
@Beamsville Beavers (Jacob) :SC: acquire Jesse Winker from @Andre - LA Smog :LA: in exchange for Jameson Taillon and $8.00 of 2023 Griffey Coin.
@Chicago Communists (Keith) :CHI: acquire Luis Garcia from @Amherst Minutemen :AMH: in exchange for prospect Owen Caissie and $7.00 of 2023 Griffey Coin.
@Beamsville Beavers (Jacob) :SC: acquire Kolton Wong from @Chicago Communists (Keith) :CHI: in exchange for $4.00 of 2023 Griffey Coin.
@Toronto Canucks :TOR: acquire Justin Steele from @Chicago Communists (Keith) :CHI: in exchange for $3.00 of 2023 Griffey Coin.
7/19/2022 9:36 AM
@Oklahoma Schooners (Matt) :OK: acquire Wilmer Flores from @McKinney Bay Knights (Ashtin) :MCK: in exchange for Eduardo Escobar and prospect Pedro Pineda.
7/19/2022 2:56 PM
@Fort Worth Fire (Austin) :FW: acquire Alex Bregman and a 2024 4th Rd OP (MCK) from @McKinney Bay Knights (Ashtin) :MCK: in exchange for Cody Bellinger, prospect Anthony Volpe, a 2023 3rd Rd OP (FW), and a 2024 2nd Rd OP (FW).
7/19/2022 4:33 PM
@Fort Worth Fire (Austin) :FW: acquire Yu Darvish, Jack Flaherty, prospect Bryce Jarvis, prospect Aaron Sabato, and prospect Austin Wells from @Zach :NO: in exchange for Walker Buehler ($8.50 retained), Sixto Sanchez, Cole Winn, and $5.00 of 2023 Griffey Coin.
7/19/2022 7:58 PM
@Chicago Communists (Keith) :CHI:  acquire Brandon Nimmo from @Andre - LA Smog :LA:  in exchange for $20.00 of 2023 Griffey Coin.
7/20/2022 1:37 PM
@Halton Huskies (Brendan) :HAL: acquire Nick Pivetta, Nathan Eovaldi, and a 2023 3rd Rd OP (LA) from @Andre - LA Smog :LA: in exchange for Maximo Castillo, David Peralta, prospect Wilman Diaz, prospect Roderick Arias, and $29.00 of 2023 Griffey Coin.
7/20/2022 2:27 PM
@Fort Worth Fire (Austin) :FW: acquire Johnny Cueto from @McKinney Bay Knights (Ashtin) :MCK: in exchange for Dean Kremer and prospect Blaze Jordan.
7/21/2022 10:49 AM
@Toronto Canucks :TOR: acquire Chris Flexen and Marco Gonzales from @Amherst Minutemen :AMH: in exchange for $5.00 of 2023 Griffey Coin.
7/21/2022 2:22 PM
@Brooklyn Brewers (Nic) :BRK: acquire Taylor Ward, Josh Rojas, and a 2023 3rd Rd OP (FW) from @McKinney Bay Knights (Ashtin) :MCK: in exchange for Evan Longoria, Ramon Urias, Garrett Crochet, prospect Michael Toglia, and a 2023 2nd Rd OP (BRK). 
@Andre - LA Smog :LA: acquire rights to T4 Free Agent Clayton Kershaw from the @Chicago Communists (Keith) :CHI: in exchange for $8.00 of 2023 Griffey Coin.
1/25/2023 11:30 AM
@Andre - LA Smog :LA: acquire rights to T3 Free Agents Pablo Lopez and Tony Gonsolin  from @Fort Worth Fire (Austin) :FW: in exchange for $15.00 of 2023 Griffey Coin.
1/25/2023 11:44 AM
@Toronto Canucks :TOR: acquire Joe Ryan, prospect Josh Jung, and rights to T3 Free Agent Pablo Lopez from @Andre - LA Smog :LA: in exchange for $10.00 of 2023 Griffey Coin
@Toronto Canucks :TOR: acquire CJ Abrams, and Tyler Mahle  from @Amherst Minutemen :AMH: in exchange for Luis Castillo, and rights to T4 Free Agent Marco Gonzales.
@Andre - LA Smog :LA: acquire rights to Dustin May from @Amherst Minutemen :AMH: in exchange for nothing
1/25/2023 11:52 AM
@Amherst Minutemen :AMH: acquire Alex Bregman from @Fort Worth Fire (Austin) :FW: in exchange for prospects Robert Hassell III and Zack Gelof, and $10.00 of 2023 Griffey Coin.
@Oklahoma Schooners (Matt) :OK: acquire rights to T4 Free Agent Sean Manaea from @Fort Worth Fire (Austin) :FW: in exchange for the rights to T4 Free Agent Kyle Gibson and $2.00 of 2023 Griffey Coin.
1/25/2023 2:44 PM
@Chicago Communists (Keith) :CHI: acquire Yu Darvish from @Fort Worth Fire (Austin) :FW: in exchange for prospect Oswald Peraza and $15.00 2023 Griffey Coin. In addition Fort Worth retains $0.03 on Yu Darvish.
1/25/2023 3:12 PM
@McKinney Bay Knights (Ashtin) :MCK: acquire Zach Eflin from @Beamsville Beavers (Jacob) :SC: in exchange for rights to T2 Free Agent Cody Bellinger and $21.00 of 2023 Griffey Coin.
1/25/2023 4:01 PM
@Amherst Minutemen :AMH: acquire Marcus Semien, Will Smith, and the rights to T2 Free Agent  Giancarlo Stanton from @Fort Worth Fire (Austin)  :FW: in exchange for prospects Noelvi Marte and Brennan Davis.
1/26/2023 10:17 AM
@Wachusett Mountaineers :COL: acquire Ke'Bryan Hayes and Trevor Larnach from @Amherst Minutemen :AMH: in exchange for $30.00 of 2023 Griffey Coin.
2/1/2023 8:17 AM
@Andre - LA Smog :LA: acquire prospect James Triantos and rights to T4 Free Agent Noah Syndergaard from @Beamsville Beavers (Jacob) :SC: in exchange for prospects Nick Yorke and Travis Swaggerty.
@Fort Worth Fire (Austin) :FW: acquire Chris Taylor, Joey Votto, and prospect Andy Pages from @Andre - LA Smog :LA:in exchange for no return.
@Fort Worth Fire (Austin) :FW: acquire Matt Brash and prospect Dustin Harris from @Amherst Minutemen :AMH: in exchange for no return.
2/1/2023 9:07 PM
@McKinney Bay Knights (Ashtin) :MCK: acquire Gavin Lux from @Fort Worth Fire (Austin) :FW: in exchange for Josh Lowe, Nick Madrigal, and $30.00 of 2023 Griffey Coin.
2/2/2023 8:29 AM
@McKinney Bay Knights (Ashtin) :MCK: acquire prospect Kahlil Watson from @Wachusett Mountaineers :COL: in exchange for $3.00 of 2023 Griffey Coin.
2/2/2023 8:44 AM
@McKinney Bay Knights (Ashtin) :MCK: acquire prospect Cole Wilcox from @Princeton Anglers (GM - Dain) @Sanger Vientos - Brandon :PRI:in exchange for $2.00 of 2023 Griffey Coin.
2/2/2023 8:57 AM
@McKinney Bay Knights (Ashtin) :MCK: acquire JJ Bleday from @Andre - LA Smog :LA:in exchange for $1.00 of 2023 Griffey Coin and a photo of Blaze Jordan.
2/2/2023 8:41 PM
@Zach :NO: acquire George Springer, Jack Flaherty, prospect Dustin Harris, and $15.00 of 2024 Griffey Coin from @Fort Worth Fire (Austin) :FW: in exchange for Gerrit Cole, Walker Buehler, and $30.00 of 2023 Griffey Coin. In addition Fort Worth retains 50% of George Springer's contract ($6.00) and 50% of Jack Flaherty's contract ($7.50).
2/2/2023 9:29 PM
@Amherst Minutemen :AMH: acquire rights to T1 Free Agents Francisco Lindor, Brandon Drury, J.D. Martinez, and Andrew Benintendi from @Fort Worth Fire (Austin) :FW: in exchange for rights to T4 Free Agents Colin Holderman and Jackson Stephens, $5.00 of 2023 Griffey Coin, and $10.00 of 2024 Griffey Coin. 
2/6/2023 11:57 AM
@Halton Huskies (Brendan) :HAL: acquire "prospect" Nate Pearson from @Princeton Anglers (GM - Dain) @Sanger Vientos - Brandon :PRI:in exchange for $2.00 of 2023 Griffey Coin.
@Amherst Minutemen :AMH: acquire right to T1 Free Agents Adolis Garcia and Danny Vogelbach and a 2023 22nd Rd OP (AMH) #300 overall from @Andre - LA Smog :LA: in exchange for rights to T4 Free Agents Joe Kelly and Joel Payamps and a 2023 2nd Rd OP (NO) #21 overall.
2/8/2023 4:08 PM
@Amherst Minutemen :AMH: acquire rights to T1 Free Agent Yandy Diaz, a 2023 21st Rd OP (LA), and $11.00 of 2023 Griffey Coin from @Andre - LA Smog :LA:in exchange for rights to T4 Free Agent Jonathan Davis and a 2023 1st Rd OP (AMH) #6 overall.
2/8/2023 4:29 PM
@McKinney Bay Knights (Ashtin) :MCK: acquire a 2023 1st Rd OP (NO) and a 2023 22nd Rd OP (COL) from @Wachusett Mountaineers :COL:in exchange for a 2023 2nd Rd OP (FW) and a 2023 2nd Rd OP (SC).
2/9/2023 3:17 PM
@Oklahoma Schooners (Matt) :OK: acquire rights to T1 Free Agents Nathaniel Lowe and Tommy La Stella from @Princeton Anglers (GM - Dain) @Sanger Vientos - Brandon :PRI: in exchange for $4.00 of 2023 Griffey Coin.
@Oklahoma Schooners (Matt) :OK: acquire rights to T1 Free Agent Seth Brown and prospect Austin Hendrick from @Wachusett Mountaineers :COL: in exchange for $1.00 of 2023 Griffey Coin.
2/10/2023 12:05 PM
@Beamsville Beavers (Jacob) :SC: acquire Aaron Ashby from @Andre - LA Smog :LA: in exchange for $5.00 of 2023 Griffey Coin.
2/15/2023 6:50 PM
@Zach :NO: acquire Randy Arozarena ($2.60 Retained), Alex Verdugo ($2.60 Retained), and prospect Luken Baker from @Brooklyn Brewers (Nic) :BRK:in exchange for prospect Sixto Sanchez and $45 of 2023 Griffey Coin. 
2/16/2023 1:14 PM
@Amherst Minutemen :AMH: acquire $20.00 of 2023 Griffey Coin from @Zach :NO:in exchange for $24.00 of 2024 Griffey Coin.
@Beamsville Beavers (Jacob) :SC: acquire $20.00 of 2023 Griffey Coin from @McKinney Bay Knights (Ashtin) :MCK: in exchange for $25.00 of 2024 Griffey Coin.
2/16/2023 2:50 PM
@McKinney Bay Knights (Ashtin) :MCK: acquire prospects Asa Lacy, Matt Allan, and Ty Madden from @typo21 :DAL:in exchange for $8.00 of 2024 Griffey Coin. 
@McKinney Bay Knights (Ashtin) :MCK: acquire Carson Kelly, Clint "Jackson" Frazier, Marcel Ozuna, and MJ Melendez from @Brooklyn Brewers (Nic) :BRK:in exchange for prospects BLAZE JORDAN , Ty Madden, Matt Allan, a 2023 2nd Rd OP (MCK) #18 OVR, a 2023 2nd Rd OP (BRK) #28 OVR, and $9.00 of 2023 Griffey Coin.
2/16/2023 3:10 PM
@Beamsville Beavers (Jacob) :SC: acquire rights to T3 Free Agent Yoan Moncada, T4 Free Agent Lance McCullers Jr., prospects Nolan Jones and Simeon Woods-Ricahrdson from @typo21 :DAL: in exchange for Josh Smith, rights to T4 Free Agent Kris Bryant, and $20 of 2023 Griffey Coin.
2/16/2023 3:43 PM
@Fort Worth Fire (Austin) :FW: acquire Clint "Jackson" Frazier, Carson Kelly, and $10.00 of 2023 Griffey Coin from @McKinney Bay Knights (Ashtin) :MCK: in excahnge for no return.
2/17/2023 10:01 AM
@McKinney Bay Knights (Ashtin) :MCK: acquire Ryan Mountcastle and prospect Heliot Ramos from @Princeton Anglers (GM - Dain) @Sanger Vientos - Brandon :PRI: in exchange for JJ Bleday and $20.00 of 2023 Griffey Coin.
2/17/2023 11:12 AM
@Chicago Communists (Keith) :CHI: acquire prospects Heriberto Hernandez and Clayton Beeter from @Andre - LA Smog :LA:in exchange for $1.00 of 2023 Griffey Coin.
2/19/2023 12:13 AM
@Beamsville Beavers (Jacob) :SC: acquire prospect Kristian Robinson from @Amherst Minutemen :AMH: in exchange for $6.00 of 2023 Griffey Coin.
2/22/2023 8:00 AM
@Amherst Minutemen :AMH: acquire rights to T1 Free Agent Jonah Heim from @Wachusett Mountaineers :COL: in exchange for $1.00 of 2023 Griffey Coin and the rights to T4 Free Agent Sam Haggerty.
3/22/2023 10:36 AM
@typo21 :DAL: acquire Michael Kopech and prospect Owen Caissie from @Amherst Minutemen :AMH: in exchange for prospect Kyle Manzardo. 
3/22/2023 12:42 PM
@McKinney Bay Knights (Ashtin) :MCK: acquire a 2023 1st Rd OP (COL) #1 OVR from @Wachusett Mountaineers :COL: in exchange for a 2023 1st RD OP (MCK) #4 OVR and $20.00 of 2024 Griffey Coin.
3/23/2023 4:07 PM
@McKinney Bay Knights (Ashtin) :MCK: acquire Martin Perez and Jon Gray from @Fort Worth Fire (Austin) :FW: in exchange for Shintaro Fujinami and Gavin Lux.
3/29/2023 8:20 AM
@Beamsville Beavers (Jacob) :SC: acquire Michael Massey and a 2024 5th Rd OP (AMH) from @Amherst Minutemen :AMH:in exchange for a 2024 3rd Rd OP (SC) and $2.00 of 2024 Griffey Coin.
3/31/2023 5:52 PM
@McKinney Bay Knights (Ashtin) :MCK: acquire Gerrit Cole and Adolis Garcia from @Fort Worth Fire (Austin) :FW: in exchange for Clarke Schmidt, prospect Anthony Volpe, a 2024 2nd Rd OP (FW), and $17 of 2024 Griffey Coin.
@Fort Worth Fire (Austin) :FW: acquire Eury Perez from @Andre - LA Smog :LA:in exchange for prospect Gavin Stone, prospect Cam Collier, prospect Andy Pages, a 2024 2nd Rd OP (FW) and $5.00 of 2024 Griffey Coin.
3/31/2023 10:00 PM
@typo21 :DAL: acquire prospect Grayson Rodriguez and Spencer Torkelson from @Amherst Minutemen :AMH: in exchange for prospect Jasson Dominguez
4/4/2023 8:15 AM
@Andre - LA Smog :LA: acquire Noah Syndergaard and Walker Buehler from @Fort Worth Fire (Austin) :FW: in exchange for Alex Kirilloff, prospect Tanner Bibee, and $5.00 of 2024 Griffey Coin.
4/4/2023 10:02 AM
@Andre - LA Smog :LA: acquire Tanner Houck (DFA) from @typo21 :DAL: in exchange for $5.00 of 2024 Griffey Coin.
4/6/2023 11:10 AM
@Amherst Minutemen :AMH: acquire Nestor Cortes Jr. from @Beamsville Beavers (Jacob) :SC: in exchange for Brandon Lowe and $7.00 of 2024 Griffey Coin.
4/7/2023 10:37 AM
@McKinney Bay Knights (Ashtin) :MCK: acquire Brian Anderson from @Fort Worth Fire (Austin) :FW: in exchange Luis Rengifo and prospect Austin Martin.
4/10/2023 8:33 AM
@Halton Huskies (Brendan) :HAL: acquire Kris Bubic (DFA) from @Beamsville Beavers (Jacob) :SC: in exchange for $5.00 of 2024 Griffey Coin. 
4/11/2023 4:53 PM
@Andre - LA Smog :LA: acquire Rowdy Tellez from @typo21 :DAL: in exchange for $3.00 of 2024 Griffey Coin.
@McKinney Bay Knights (Ashtin) :MCK: acquire Lucas Giolito from @Wachusett Mountaineers :COL: in exchange for Taijuan Walker and $5.00 of 2024 Griffey Coin.
4/13/2023 3:43 PM
@Beamsville Beavers (Jacob) :SC: acquire Nick Pivetta and Eduardo Rodriguez from @Wachusett Mountaineers :COL: in exchange for $8.00 of 2024 Griffey Coin.
4/20/2023 8:24 AM
@Chicago Communists (Keith) :CHI: acquire Marcus Stroman from @Wachusett Mountaineers :COL:in exchange for  Stroman for Jose Suarez and $10.00 of 2023 Griffey Coin.
4/20/2023 9:04 AM
@Amherst Minutemen :AMH: acquire Jordan Lyles from @Fort Worth Fire (Austin) :FW:  in exchange for Jalen Beeks and $2.00 of 2024 Griffey Coin. (retroactive to 4/17)
4/20/2023 11:40 AM
@typo21 :DAL: acquire Nathaniel Lowe from @Zach :NO: in exchange for $5.00 of 2024 Griffey Coin.
4/24/2023 8:44 AM
@Fort Worth Fire (Austin) :FW: acquire David Villar and $3.00 of 2024 Griffey Coin from @McKinney Bay Knights (Ashtin) :MCK: in exchange for Jhony Brito.
4/25/2023 10:42 AM
@Toronto Canucks :TOR: acquire Anthony Santander from @Andre - LA Smog :LA: in exchange for $2.00 of 2024 Griffey Coin.
4/27/2023 10:23 AM
@Andre - LA Smog :LA: acquire prospect Tanner Bibee from @Fort Worth Fire (Austin) :FW: in exchange for prospect Cam Collier.
4/27/2023 10:41 AM
@Fort Worth Fire (Austin) :FW: acquire Brenton Doyle from @McKinney Bay Knights (Ashtin) :MCK: in exchange for $2.00 of 2024 Griffey Coin.
5/11/2023 9:12 AM
@Beamsville Beavers (Jacob) :SC: acquire Tanner Houck from @Andre - LA Smog :LA: in exchange for $5.00 of 2024 Griffey Coin.
5/15/2023 1:15 PM
@Andre - LA Smog :LA: acquire David Peterson from @Beamsville Beavers (Jacob) :SC:in exchange for $1.00 of 2024 Griffey Coin.
@Chicago Communists (Keith) :CHI: acquire Nathan Eovaldi and Matt Mervis from @Halton Huskies (Brendan) :HAL: in exchange for Alek Manoah and $5.00 of 2024 Griffey Coin. 
5/17/2023 12:57 PM
@McKinney Bay Knights (Ashtin) :MCK: acquire Nick Senzel from @Wachusett Mountaineers :COL: in exchange for Brian Anderson and $4.00 of 2024 Griffey Coin.
5/17/2023 3:21 PM
@Brooklyn Brewers (Nic) :BRK: acquire Mark Vientos from @Amherst Minutemen :AMH:in exchange for $5.00 of 2024 Griffey Coin.
5/17/2023 10:34 PM
@Chicago Communists (Keith) :CHI: acquire Max Muncy and Corbin Burnes from @Andre - LA Smog :LA: in exchange for Eugenio Suarez, prospects Elly de la Cruz, Jackson Merrill, and Tink Hence and $18.00 of 2024 Griffey Coin. LA retains $2.50 on Corbin Burnes.
5/18/2023 1:00 PM
@Oklahoma Schooners (Matt) :OK: acquire Bryce Harper from @Andre - LA Smog :LA:in exchange for prospects Jordan Lawlar, Taj Bradley, Adael Amador, and $10.00 of 2024 Griffey Coin. LA retains $2.25 on Bryce Harper. 
@Amherst Minutemen :AMH: acquire Masataka Yoshida from @Wachusett Mountaineers :COL: in exchange for prospects Kyle Manzardo and Miguel Bleis.
5/18/2023 7:33 PM
@McKinney Bay Knights (Ashtin) :MCK: acquire Tyler Glasnow and Taj Bradley from @Andre - LA Smog :LA: in exchange for prospects Curtis Mead, Tyler Soderstrom, Heston Kjerstad, Brayan Rocchio, and $15.00 of 2024 Griffey Coin.
5/23/2023 2:45 PM
@Amherst Minutemen :AMH: acquire LaMonte Wade Jr. from @Andre - LA Smog :LA:in exchange for Harold Ramirez and $5.00 of 2024 Griffey Coin.
@Beamsville Beavers (Jacob) :SC: acquire Isaac Parades from @Wachusett Mountaineers :COL:in exchange for $4.00 of 2024 Griffey Coin.
@Halton Huskies (Brendan) :HAL: acquire Jeff McNeil from @Andre - LA Smog :LA:in exchange for $5.00 of 2024 Griffey Coin.
@Brooklyn Brewers (Nic) :BRK: acquire Chris Sale from @Andre - LA Smog :LA:in exchange for a 2024 3rd Rd OP (BRK), a 2025 1st Rd OP (BRK), and $18.00 of 2024 Griffey Coin.
5/23/2023 3:24 PM
@Andre - LA Smog :LA: acquire Ian Happ from @typo21 :DAL: in exchange for prospect Andy Pages.
5/26/2023 11:28 PM
@Brooklyn Brewers (Nic) :BRK: acquire Marcel Ozuna from @McKinney Bay Knights (Ashtin) :MCK: in exchange for $5.00 of 2024 Griffey Coin.
5/28/2023 7:38 PM
@Beamsville Beavers (Jacob) :SC: acquire Garrett Whitlock from @Fort Worth Fire (Austin) :FW: in exchange for $6.00 of 2024 Griffey Coin.
@Princeton Anglers (GM - Dain) :PRI: acquire Ryan Mountcastle, Michael Soroka, and Jared Shuster from @McKinney Bay Knights (Ashtin) :MCK: in exchange for Pete Alonso.
5/30/2023 10:07 AM
@McKinney Bay Knights (Ashtin) :MCK: acquire Brandon Drury (DFA) from @Halton Huskies (Brendan) :HAL: in exchange for $1.00 of 2024 Griffey Coin.
6/1/2023 1:42 PM
@Beamsville Beavers (Jacob) :SC: acquire Chris Sale from @Brooklyn Brewers (Nic) :BRK: in exchange for prospects Spencer Jones and Jace Jung, a 2025 1st OP (SC) and $8.00 of 2024 Griffey Coin. 
6/1/2023 2:40 PM
@McKinney Bay Knights (Ashtin) :MCK: acquire TIER ONE from @Fort Worth Fire (Austin) :FW: in exchange for Ryan Jeffers,  Alek Thomas and $5.00 of 2024 Griffey Coin. 
6/14/2023 11:39 AM
@Amherst Minutemen :AMH: acquire Dylan Cease, Mike Clevinger, Michael Kopech, and Hellen's Kid (Mitch Keller) from @Sanger Vientos - Brandon :DAL: in exchange for Mackenzie Gore, prospect Zach Neto and prospect EVAN CARTER. Interim Interim's retain $4.05 (50%) of Dylan Cease contract.
6/14/2023 11:48 AM
@Beamsville Beavers (Jacob) :SC: acquire Dean Kremer and Kyle Bradish from @Amherst Minutemen :AMH: in exchange for $6.00 of 2024 Griffey Coin.
6/14/2023 3:02 PM
@Zach :NO: acquire Spencer Torkelson from @Sanger Vientos - Brandon :DAL:in exchange for Andrew Vaughn.
6/14/2023 3:12 PM
@Brooklyn Brewers (Nic) :BRK: acquire Jose Ramirez ,prospect Brandon Pfaadt, and $5.00 of 2024 Griffey Coin from @Sanger Vientos - Brandon :DAL: in exchange for prospects Esteury Ruiz and Ezequiel Duran. DAL retains $6.50 on Jose Ramirez
@Brooklyn Brewers (Nic) :BRK: acquire prospect Bryce Jarvis and a 2024 3rd Rd OP (FW) from @Fort Worth Fire (Austin) :FW: in exchange for Edward Cabrera, prospect Dax Fulton, and $2.00 of 2024 Griffey Coin.
@Fort Worth Fire (Austin) :FW: acquire Jesus Luzardo from @Sanger Vientos - Brandon :DAL: in exchange for Nick Madrigal, prospect Owen White and prospect Oswald Peraza. DAL retains $3.50 on Jesus Luzardo. 
6/20/2023 8:33 AM
@Beamsville Beavers (Jacob) :SC: acquire Chris Bassitt and $5.00 of 2024 Griffey Coin from @Brooklyn Brewers (Nic) :BRK: in exchange forEzequiel Duran, prospect Colt Keith, prospect Dylan Lesko, and prospect Edouard Julien.
@Sanger Vientos - Brandon :DAL: acquire prospect Tyler Soderstrom from @Andre - LA Smog :LA:in exchange for Nathaniel Lowe.
@Zach :NO: acquire Sean Murphy from @Sanger Vientos - Brandon :DAL: in exchange for Alejandro Kirk and prospect Cole Winn.
6/20/2023 8:44 AM
@Brooklyn Brewers (Nic) :BRK: acquire Ronny Mauricio from @Sanger Vientos - Brandon :DAL: in exchange for $5.00 of 2024 Griffey Coin.
@Halton Huskies (Brendan) :HAL: acquire Shane Bieber and Aaron Civale from @Brooklyn Brewers (Nic) :BRK: in exchange for Seth Lugo, prospect Spencer Steer, prospect Logan O'Hoppe and $5.00 of 2024 Griffey Coin. BRK retains $2.50 on Shane Bieber. 
@Wachusett Mountaineers :COL: acquire Andres Gimenez and prospect Kahlil Watson from @McKinney Bay Knights (Ashtin) :MCK:in exchange for $10.00 of 2024 Griffey Coin.
@McKinney Bay Knights (Ashtin) :MCK: acquire Justin Turner from @Andre - LA Smog :LA:in exchange for $14.00 of 2024 Griffey Coin.
@Princeton Anglers (GM - Dain) :PRI: acquire AJ Smith-Shawver and Jorge Soler from @Andre - LA Smog :LA: in exchange for prospect Michael Busch and $5.00 of 2024 Griffey Coin.
@Toronto Canucks :TOR: acquire Bailey Ober from @Fort Worth Fire (Austin) :FW: in exchange for $12.00 of 2024 Griffey Coin.
6/27/2023 10:32 AM
@Zach :NO: acquire Blake Snell and $3.00 of 2024 Griffey Coin from @Sanger Vientos - Brandon :DAL: in exchange for Ha-Seong Kim and Spencer Torkelson. Dallas retains $4.07 on Blake Snell.
@Andre - LA Smog :LA: acquire Griffin Canning from @Chicago Communists (Keith) :CHI: in exchange for $7.00 of 2024 Griffey Coin.
6/27/2023 10:43 AM
@Andre - LA Smog :LA: acquire Braxton Garrett from @Wachusett Mountaineers :COL: in exchange for $8.00 of 2024 Griffey Coin.
6/29/2023 3:30 PM
@Chicago Communists (Keith) :CHI: acquire Francisco Lindor from @Sanger Vientos - Brandon :DAL: in exchange for prospect Alex Ramirez, prospect Bryan Ramos, a 2024 1st Rd OP (CHI) and $15.00 of 2024 Griffey Coin. DAL retains $3.00 on Francisco Lindor. 
@Sanger Vientos - Brandon :DAL: acquire Patrick Corbin from @Toronto Canucks :TOR: in exchange for future considerations that will never be considered (aka nothing).
6/29/2023 3:38 PM
@Beamsville Beavers (Jacob) :SC: acquire Whit Merrifield from @Oklahoma Schooners (Matt) :OK: in exchange for Jose Urquidy.
6/30/2023 11:59 AM
@Sanger Vientos - Brandon :DAL: acquire Dalton Varsho and Anthony DeSclafani from @Oklahoma Schooners (Matt) :OK:in exchange for prospect Hunter Brown and $5.00 of 2024 Griffey Coin.
6/30/2023 12:48 PM
@Brooklyn Brewers (Nic) :BRK: acquire Alex Cobb, and Carlos Rodon from @Oklahoma Schooners (Matt) :OK: in exchange for Luis Garcia, J.P. France and $5.00 of 2024 Griffey Coin.
7/6/2023 8:33 AM
@Zach :NO: acquire some guy with the last name Sanchez from @Fort Worth Fire (Austin) :FW: in exchange for $1.00 of 2024 Griffey Coin. 
@McKinney Bay Knights (Ashtin) :MCK: acquire Lance Lynn from @Oklahoma Schooners (Matt) :OK: in exchange for David Peterson, prospect DL Hall and prospect Pedro Pineda.
7/6/2023 4:09 PM
@Amherst Minutemen :AMH: acquire Joey Votto from @Fort Worth Fire (Austin) :FW: in exchange for $1.00 of 2024 Griffey Coin.
7/13/2023 10:40 AM
@Andre - LA Smog  :LA: acquire prospect Evan "The Kid" Carter from @Sanger Vientos - Brandon :DAL: in exchange for prospects Tink Hence, Elijah Green, Daniel Espoisto, and Bo Naylor.
@Toronto Canucks :TOR: acquire Byron Buxton from @Amherst Minutemen :AMH: in exchange for Cole's pain to go away of having him on his roster (nothing). Amherst retains $2.75 on Buxton.
@Brooklyn Brewers (Nic) :BRK: acquire Amed Rosario (DFA) from @Zach :NO: in exchange for $1.00 of 2024 Griffey Coin. 
7/13/2023 6:56 PM
@Amherst Minutemen :AMH: acquire Wade Miley from @Fort Worth Fire (Austin) :FW: in exchange for a 2024 3rd Rd OP (AMH).
7/13/2023 8:24 PM
@Toronto Canucks :TOR: acquire Carlos Correa from @Fort Worth Fire (Austin) :FW: in exchange for a 2024 2nd Rd OP (TOR). Fort Worth retains $1.50 on Correa.
7/14/2023 7:29 AM
@Brooklyn Brewers (Nic) :BRK: acquire Yu Darvish and prospect Ryne Nelson from @Chicago Communists (Keith) :CHI: in exchange for Alex Cobb.
2/9/2024 9:46 AM
@Beamsville Beavers (Jacob) :SC: acquire prospect James Triantos and a 2024 2nd Rd OP ( :LA: ; #20 Overall) from @Andre - LA Smog :LA: in exchange for a 2024 2nd Rd OP ( :SC: ; #23 Overall) and $12.00 of 2024 Griffey Coin.
2/10/2024 11:15 AM
@Oklahoma Schooners (Matt) :OK: acquire FA rights to Marcus Semien from @Amherst Minutemen :AMH: in exchange for FA rights to Justin Turner and $5.00 of 2024 Griffey Coin.
2/15/2024 4:31 PM
@McKinney Bay Knights (Ashtin) :MCK: acquire prospect Kahlil Watson from @Wachusett Mountaineers :COL: in exchange for $1.00 of 2025 Griffey Coin.
@Toronto Canucks :TOR: acquire prospect Dustin Harris from @Wachusett Mountaineers  :COL: in exchange for $1.00 of 2024 Griffey Coin.
@Toronto Canucks :TOR: acquire prospect George Valera from @McKinney Bay Knights (Ashtin) :MCK: in exchange for $1.00 of 2025 Griffey Coin.
2/16/2024 10:40 AM
@Andre - LA Smog :LA: acquire Starling Marte and $12.00 of 2024 Griffey Coin from @Toronto Canucks :TOR: in exchange for future considerations.
2/19/2024 10:39 AM
@Zach :NO: acquire Matthew Liberatore from @Amherst Minutemen :AMH: in exchange for rights to Xander Bogaerts
2/19/2024 10:45 AM
@Sanger Vientos - Brandon :DAL:  acquire Ezequiel Tovar, prospect Brayan Rocchio, prospect Heston Kjerstad, and Tony Gonsolin from @Andre - LA Smog  :LA: in exchange for $70.00 of 2024 Griffey Coin. 
2/19/2024 10:59 AM
@Halton Huskies (Brendan) :HAL: acquire Chris Bassitt and a 2024 2nd Rnd OP (20th Overall LA) from @Beamsville Beavers (Jacob) :BEA: for prospect Trey Sweeney, prospect Ivan Melendez, and $10.00 of 2024 Griffey Coin.
3/14/2024 7:47 PM
@Amherst Minutemen :AMH: acquire Brayan Bello and prospect Colt Emerson from @Andre - LA Smog :LA: in exchange for prospect Emmanuel Rodriguez
@Amherst Minutemen :AMH: acquire the FA rights to Yu Darvish from @Brooklyn Brewers (Nic) :BRK: in exchange for the FA rights to Nestor Cortes
@Amherst Minutemen :AMH: acquire the FA rights to Nick Pivetta from @Fort Worth Fire (Austin) :FW: in exhange for $1 GC and the FA rights to Jarred Kelenic
@Amherst Minutemen :AMH: acquire the FA rights to Christian Yelich from @Beamsville Beavers (Jacob) :SC: in exchange for 2024 1st Round OP (1-13)
3/14/2024 7:56 PM
@Wachusett Mountaineers :COL: acquire Mike Trout and prospects Heriberto Hernandez and Benny Montgomery from @Chicago Communists (Keith) :CHI: in exchange for prospects Mick Abel and Miguel Bleis
3/15/2024 8:38 AM
@Amherst Minutemen :AMH: acquire FA rights to Kutter Crawford from @McKinney Bay Knights (Ashtin) :MCK: in exchange for $2 GC 
3/19/2024 1:58 PM
@Wachusett Mountaineers :COL: acquire the FA rights to Mitch Keller from @Amherst Minutemen :AMH: in exchange for $4 GC
3/19/2024 2:55 PM
@Toronto Canucks :TOR: acquire Dylan Cease, FA rights to Yu Darvish, and prospects Arjun Nimmala and David Festa from @Amherst Minutemen :AMH: in exchange for prospects Graham Pauley and Emerson Hancock
3/19/2024 7:06 PM
@Beamsville Beavers (Jacob) :SC: acquire the FA rights to Nolan Schanuel from @Amherst Minutemen :AMH: in exchange for $1 GC
3/21/2024 3:41 PM
@Toronto Canucks :TOR: acquire Kyle Schwarber and $5.00 of 2025 Griffey Coin from @Oklahoma Schooners (Matt) :OK: in exchange for Cristian Javier.
3/25/2024 2:57 PM
@Fort Worth Fire (Austin) :FW: acquire Aaron Civale and Yainer Diaz from @Halton Huskies (Brendan) :HAL: in exchange for William Contreras and $7.00 of 2025 Griffey Coin.
3/26/2024 1:24 PM
@McKinney Bay Knights (Ashtin) :MCK: acquire Aaron Civale from @Fort Worth Fire (Austin) :FW:  in exchange for Bowden Francis and $25.00 of 2025 Griffey Coin.
3/26/2024 7:52 PM
@Beamsville Beavers (Jacob) :SC: acquire Brady Singer (DFA) from @Andre - LA Smog :LA: in exchange for $6.00 of 2025 Griffey coin. LA retains $2.50 on Brady Singer. 
3/26/2024 9:50 PM
@Brooklyn Brewers (Nic) :BRK: acquire Chris Sale and prospect Thomas Saggese from @Beamsville Beavers (Jacob) :SC: in exchange for Blake Snell, prospect Ty Madden and $10.00 of 2025 Griffey Coin.
3/27/2024 1:48 PM
@Halton Huskies (Brendan) :HAL: acquire Mike Trout from @Wachusett Mountaineers :COL: in exchange for Jake Burger, prospect Hurston Waldrep, prospect Nobel Meyer, prospect Ethan Salas, and $20.00 of 2025 Griffey Coin. Colorado retains 50% of Mike Trout. 
3/28/2024 8:13 AM
@Fort Worth Fire (Austin) :FW: acquire AJ Puk from @Halton Huskies (Brendan) :HAL: in exchange for $1.00 of 2025 Griffey Coin.
3/28/2024 8:24 AM
@Chicago Communists (Keith) :CHI: acquire Grayson Rodriguez from @Sanger Vientos - Brandon :DAL: in exchange for Isaac Paredes, Tyler Anderson, and prospect Jacob Misiorowski.
3/31/2024 5:33 AM
@Amherst Minutemen :AMH: acquire Thairo Estrada from @Wachusett Mountaineers :COL: in exchange for no return.
@Beamsville Beavers (Jacob) :SC: acquire Josh Bell from @Chicago Communists (Keith) :CHI: in exchange for $3 GC.
3/31/2024 8:44 AM
@McKinney Bay Knights (Ashtin) :MCK: acquire Starling Marte (DFA) from @Andre - LA Smog :LA: in exchange for $1.00 of 2025 Griffey Coin
4/1/2024 10:52 AM
@Sanger Vientos - Brandon :DAL: acquire Cedric Mullins from @Amherst Minutemen :AMH: in exchange for $5 GC.
4/3/2024 8:58 PM
@Sanger Vientos - Brandon :DAL: acquire prospect Marco Luciano from @McKinney Bay Knights (Ashtin) :MCK: in exchange for prospect Heston Kjerstad.
4/5/2024 9:22 AM
@Amherst Minutemen :AMH: acquire prospect Coby Mayo from @Sanger Vientos - Brandon :DAL: in exchange for prospects Edgar Quero, George Lombard Jr and Graham Pauley
4/8/2024 12:17 PM
@Wachusett Mountaineers :COL: acquire Cavan Biggio and $2.00 of 2025 Griffey Coin from @McKinney Bay Knights (Ashtin) :MCK: in exchange for David Fry.
4/9/2024 7:53 AM
@Beamsville Beavers (Jacob) :SC: acquire Trea Turner ($1.25 retained) and Luis Castillo ($3.25 retained) from @Amherst Minutemen :AMH: in exchange for prospect James Triantos and $17 GC
4/9/2024 10:06 AM
@Amherst Minutemen :AMH: acquire JP Crawford from @Chicago Communists (Keith) :CHI: in exchange for $3 GC
4/9/2024 1:48 PM
@Andre - LA Smog :LA: acquire Will Smith, Justin Turner and Thairo Estrada (AMH retains $0.50) and $1 GC from @Amherst Minutemen :AMH: in exchange for Jackson Merrill and Curtis Mead
4/9/2024 8:13 PM
@Brooklyn Brewers (Nic) :BRK: acquire Alex Bregman from @Amherst Minutemen :AMH: in exchange for Edouard Julien and $10 GC
4/9/2024 10:14 PM
@Sanger Vientos - Brandon :DAL: acquire Cody Bradford from @Fort Worth Fire (Austin) :FW: in exchange for prospect Kevin Alcantara.
4/10/2024 10:20 AM
@Sanger Vientos - Brandon :DAL: acquire Kyle Tucker, Pete Alonso and Charlie Blackmon from @Amherst Minutemen :AMH: in exchange for Spencer Torkelson and prospects Roman Anthony, Owen Caissie and Tink Hence
4/10/2024 2:29 PM
@McKinney Bay Knights (Ashtin) :MCK: acquire Masataka Yoshida from @Amherst Minutemen :AMH: in exchange for $2.00 of 2025 Griffey Coin.
4/12/2024 12:42 PM
@Andre - LA Smog :LA: acquire Carlos Correa and prospect Roderick Arias from @McKinney Bay Knights (Ashtin) :MCK: in exchange for prospect Jackson Jobe.
@Wachusett Mountaineers :COL: acquire Nolan Schanuel, Spencer Arrighetti, and prospect Jackson Jobe from @McKinney Bay Knights (Ashtin) :MCK: in exchange for prospect Ethan Salas.
4/12/2024 2:18 PM
@Chicago Communists (Keith) :CHI: acquire Ty France (DFA'd previously) from @Amherst Minutemen :AMH: in exchange for a 7th Rd Open Pick
@Amherst Minutemen :AMH: acquire Ty France from @Chicago Communists (Keith) :CHI: in exchange for a 6th Rd Open Pick
4/14/2024 8:31 AM
@Beamsville Beavers (Jacob) :SC: acquire Cavan Biggio from @Wachusett Mountaineers :COL: in exchange for $3 GC
4/14/2024 2:06 PM
@Sanger Vientos - Brandon :DAL: acquire Luis Severino from @Amherst Minutemen :AMH: in exchange for prospect George Lombard Jr
4/15/2024 8:44 AM
@McKinney Bay Knights (Ashtin) :MCK: acquire Ross Stripling (DFA) from @Andre - LA Smog :LA:in exchange for $1.00 of 2025 Griffey Coin. 
Condition: If Ross Stripling scores less than 8 points in his start $1.00 of 2025 Griffey Coin is returned to McKinney and Ross Stripling is DFA'd. If Ross Stripling scores 8 or more points in his start, the $1.00 of Griffey Coin stands, and Ross Stripling is still DFA'd
4/15/2024 11:59 AM
@Brooklyn Brewers (Nic) :BRK: acquire Paul Goldschmidt from @Fort Worth Fire (Austin) :FW: in exchange for $12.00 of 2025 Griffey Coin.
4/15/2024 1:12 PM
@Sanger Vientos - Brandon :DAL: acquire Frankie Montas from @Andre - LA Smog :LA: in exchange for prospect Andy Pages
4/15/2024 1:56 PM
@Sanger Vientos - Brandon :DAL: acquire Jose Butto from @Amherst Minutemen :AMH: in exchange for prospect Jackson Jaison Chourio
4/17/2024 9:12 AM
@Sanger Vientos - Brandon :DAL: acquire Salvador Perez from @Oklahoma Schooners (Matt) :OK:in exchange for Luis Campusano and $2.00 of 2025 Griffey Coin.
4/17/2024 10:45 AM
@McKinney Bay Knights (Ashtin) :MCK: acquire Frankie Montas from @Sanger Vientos - Brandon :DAL:in exchange for $5.00 of 2025 Griffey Coin.
4/17/2024 2:09 PM
@Andre - LA Smog :LA: acquire Landon Knack from @McKinney Bay Knights (Ashtin) :MCK: in exchange for $1.00 of 2025 Griffey Coin.
4/21/2024 10:33 AM
@Sanger Vientos - Brandon :DAL: acquire Mitchell Parker from @Amherst Minutemen :AMH: in exchange for $3 GC
4/22/2024 10:08 AM
@Beamsville Beavers (Jacob) :SC: acquire Amed Rosario from @Brooklyn Brewers (Nic) :BRK: in exchange for $4 GC
4/22/2024 11:13 AM
@McKinney Bay Knights (Ashtin) :MCK: acquire Zach Eflin from @Wachusett Mountaineers :COL: in exchange for $10.00 of 2025 Griffey Coin. 
4/25/2024 7:27 AM
@Beamsville Beavers (Jacob) :SC: acquire prospect Addison Barger from @Chicago Communists (Keith) :CHI: in exchange for $8 GC.
4/25/2024 8:14 AM
@Beamsville Beavers (Jacob) :SC:  acquire Michael Wacha from @Toronto Canucks :TOR:  for prospect Simeon Woods-Richardson.
4/26/2024 2:17 PM
@Fort Worth Fire (Austin) :FW: acquire Masataka Yoshida from @McKinney Bay Knights (Ashtin) :MCK: in exchange for $3.00 of 2025 Griffey Coin.
4/28/2024 11:39 AM
@Sanger Vientos - Brandon :DAL: acquire Erik Fedde from @Princeton Anglers (GM - Dain) :PRI: in exchange for Jose Butto and $5.00 of 2025 Griffey Coin.
4/29/2024 10:56 PM
@Fort Worth Fire (Austin) :FW: acquire Jameson Taillon from @Andre - LA Smog :LA: in exchange for $3.00 of 2025 Griffey Coin.
5/1/2024 9:08 AM
@Sanger Vientos - Brandon :DAL:  acquire Trevor Williams from @Andre - LA Smog :LA: in exchange for $2.00 of 2025 Griffey Coin.
5/2/2024 2:16 PM
@Toronto Canucks :TOR: acquire Cole Ragans and Ryan Ohearn from @Amherst Minutemen :AMH: in exchange for prospects Samuel Zavala, Tekoah Roby and $46 GC 
5/3/2024 2:48 PM
@Wachusett Mountaineers :COL: acquire Tyler Anderson from @Sanger Vientos - Brandon :DAL: in exchange for $4 GC.
5/3/2024 3:38 PM
Homer @Sanger Vientos - Brandon :DAL: acquire Dane Dunning from @Zach :NO:  in exchange for prospect Marco Luciano and $8 GC.
5/6/2024 8:44 AM
@Sanger Vientos - Brandon :DAL: acquire Tyler Alexander from @Chicago Communists (Keith) :CHI: in exchange for $3 GC
5/6/2024 1:36 PM
@Brooklyn Brewers (Nic) :BRK: acquire Tommy Pham from @McKinney Bay Knights (Ashtin) :MCK: in exchange for Luis Garcia.
@Andre - LA Smog :LA: acquire Josh Rojas from @Brooklyn Brewers (Nic) :BRK:in exchange for $5.00 of 2025 Griffey Coin.
5/9/2024 10:09 AM
@Chicago Communists (Keith) :CHI: acquire Danny Jansen from @Andre - LA Smog :LA: in exchange for $4.00 of 2025 Griffey Coin
5/9/2024 4:54 PM
@Beamsville Beavers (Jacob) :SC: acquire Jake Cronenworth from @Fort Worth Fire (Austin) :FW: in exchange for Carson Whisenhunt and $5 GC.
5/10/2024 11:20 PM
@Sanger Vientos - Brandon :DAL: acquire Big Maple James Paxton from @Chicago Communists (Keith) :CHI: in exchange for $6.00 of 2025 Griffey Coin
5/24/2024 1:42 PM
@Halton Huskies (Brendan) :HAL: acquire Davis Schneider from @Zach :NO: in exchange for $4 GC
5/25/2024 11:35 AM
@Beamsville Beavers (Jacob) :SC: acquire Kris Bryant from @Sanger Vientos - Brandon :DAL: in exchange for $2 GC. 
TRADE CONDITION:
DAL acquire $1.00 GC from SC if Kris Bryant plays 90 games this season. 
5/28/2024 8:30 PM
@Beamsville Beavers (Jacob) :SC: acquire Yordan Alvarez from @Princeton Anglers (GM - Dain) :PRI: in exchange for Brendan Donovan (minors), Nolan Jones (minors), prospect Kyle Teel, prospect Jacob Wilson, prospect Austin Gauthier, and $5.00 of 2025 Griffey Coin. Princeton retains $7.00 of Yordan Alvarez contract in 2026 and 2027.
5/29/2024 3:25 PM
@Wachusett Mountaineers :COL: acquire Bryce Harper from @Oklahoma Schooners (Matt) :OK: in exchange for prospects Colton Cowser, Kyle Manzardo, Brady House, Termarr Johnson, and Hurston Waldrep. Oklahoma retains $3.00 of Harpers contract in 2025.
5/30/2024 12:08 PM
@Chicago Communists (Keith) :CHI: acquire Albert Suarez from @Amherst Minutemen :AMH: in exchange for $3 GC.
5/30/2024 1:28 PM
@Halton Huskies (Brendan) :HAL: acquire Julio Rodriguez from @Amherst Minutemen :AMH: in exchange for prospects Cole Young and Noah Schultz and $25 GC. 
5/30/2024 1:37 PM
@Wachusett Mountaineers :COL: acquire Jurickson Profar and Luis Rengifo from @McKinney Bay Knights (Ashtin) :MCK: in exchange for Gavin Sheets, Brandon Marsh, Jordan Walker, prospect Zac Veen and Druw Jones.
5/30/2024 2:56 PM
@Amherst Minutemen :AMH: acquire Jarred Kelenic from @Toronto Canucks :TOR: in exchange for Elias Diaz.
Condition: If Kelenic remains on Amherst's roster until the end of the season, Toronto will send $1 GC to Amherst. 
5/30/2024 9:18 PM
@Brooklyn Brewers (Nic) :BRK: acquire Logan Webb from @Oklahoma Schooners (Matt) :OK: in exchange for Christian Encarnacion-Strand, prospect Jace Jung and prospect Dylan Lesko.
6/2/2024 9:30 AM
@Andre - LA Smog :LA: acquire Ketel Marte and Brandon Nimmo from @Zach :NO: in exchange for Adael Amador, Matt Shaw, Roderick Arias, and $11 GC.
6/3/2024 7:20 PM
@Sanger Vientos - Brandon :DAL: acquire Jacob deGrom from @McKinney Bay Knights (Ashtin) :MCK: in exchange for prospect Henry Davis and $10.00 of 2025 Griffey Coin.
6/3/2024 7:34 PM
@Brooklyn Brewers (Nic) :BRK: acquire Zach Eflin from @McKinney Bay Knights (Ashtin) :MCK: in exchange for $10.00 of 2025 Griffey Coin.
6/6/2024 8:57 AM
@Oklahoma Schooners (Matt) :OK: acquire Randy Vasquez from @Toronto Canucks :TOR: in exchange for $1.00 of 2025 Griffey Coin. 
Condition: If Vasquez DOES NOT pitch a perfect game on 6/6/2024 against the Arizona Diamondbacks, Oklahoma sends an additional $1.00 of 2025 Griffey Coin to Toronto. 
6/6/2024 1:51 PM
@Sanger Vientos - Brandon :DAL: acquire Mark Vientos and Ben Lively from @Amherst Minutemen :AMH: in exchange for prospect Jacob Misiorowski and $5 GC.
6/7/2024 10:38 AM
@Andre - LA Smog :LA: acquire Starling Marte (DFA) from @McKinney Bay Knights (Ashtin) :MCK:in exchange for Jeimer Candelario (DFA).
Condition: Andre will now vote for Trump in 2024 election
6/9/2024 9:15 AM
@McKinney Bay Knights (Ashtin) :MCK: acquire Charlie Blackmon from @Sanger Vientos :DAL: in exchange for $1.00 of 2025 Griffey Coin.
6/9/2024 9:57 AM
@Beamsville Beavers (Jacob) :SC: acquire Yu Darvish from @Zach :NO: in exchange for Brandon Woodruff, prospect Kristian Robinson, and $5.00 of 2025 Griffey Coin
@Beamsville Beavers (Jacob) :SC: acquire KeBryan Hayes from @Amherst Minutemen :AMH: in exchange for Rhys Hoskins and $3.00 of 2025 Griffey Coin
6/10/2024 9:20 PM
@Brooklyn Brewers (Nic) :BRK: acquire Charlie Blackmon from @McKinney Bay Knights (Ashtin) :MCK: in exchange for Hitler Adell and $2.00 of 2024 Griffey Coin.
6/12/2024 8:24 AM
@Andre - LA Smog :LA: acquire Ryan Mountcastle and Jeimer Candelario from @McKinney Bay Knights (Ashtin) :MCK: in exchange for prospects Andy Pages and Justyn-Henry Malloy.
6/14/2024 8:14 PM
@Wachusett Mountaineers :COL: acquire Jack Flaherty and Marcell Ozuna from @Beamsville Beavers (Jacob) :SC: in exchange for Miguel Vargas, prospect Joey Ortiz, prospect Max Clark and $45.00 of 2025 Griffey Coin.
6/14/2024 9:21 PM
@Beamsville Beavers (Jacob) :SC: acquire some random reds OF (TJ Friedl) from @Fort Worth Fire (Austin) :FW: in exchange for $6.00 of 2025 Griffey Coin. 
6/19/2024 12:41 PM
@Amherst Minutemen :AMH: acquire Mitch Garver from @Andre - LA Smog :LA: in exchange for $1 GC
6/21/2024 9:42 AM
@Beamsville Beavers (Jacob) :SC: acquire Willy Adames from @Zach :NO: in exchange for $8 GC
6/21/2024 3:15 PM
@Wachusett Mountaineers :COL: acquire Jarren Duran and Jameson Taillon from @Fort Worth Fire (Austin) :FW: in exchange for Spencer Arispaghetti, prospects Jackson Jobe and Noble Meyer and $10 of 2025 GC
6/22/2024 11:39 AM
@Amherst Minutemen :AMH: acquire Jeimer Candelario from @Andre - LA Smog :LA: in exchange for $1 GC.
6/22/2024 1:07 PM
@Beamsville Beavers (Jacob) :SC: acquire Alex Verdugo from @Zach :NO: in exchange for $8 GC
6/24/2024 9:29 AM
@Oklahoma Schooners (Matt) :OK: acquire Patrick Corbin from @Sanger Vientos - Brandon  :DAL:  in exchange for $1 GC. 
6/26/2024 10:26 AM
@Wachusett Mountaineers :COL: acquire Spencer Turnbull from @Fort Worth Fire (Austin) :FW: in exchange for $2 GC.
7/7/2024 9:49 AM
@Sanger Vientos - Brandon :DAL: acquire Brent Rooker from @Amherst Minutemen :AMH: in exchange for Brett Baty
7/13/2024 3:25 PM
@Brooklyn Brewers (Nic) :BRK: acquire Reynaldo Lopez from @Fort Worth Fire (Austin) :FW: in exchange for Prospect Saggese and $15 GC
7/15/2024 2:51 PM
@Oklahoma Schooners (Matt) :OK: acquire Byron Buxton from @Toronto Canucks :TOR: in exchange for $1 GC.
7/17/2024 9:51 AM
@Beamsville Beavers (Jacob) :SC: acquire Aaron Civale from @McKinney Bay Knights (Ashtin) :MCK: in exchange for $2.00 of 2025 Griffey Coin.
7/17/2024 10:30 AM
@Brooklyn Brewers (Nic) :BRK: acquire Robbie Ray from @McKinney Bay Knights (Ashtin) :MCK: in exchange for Jose Soriano, prospect Ronny Mauricio, and a 2025 2nd Rd Open Draft Pick (BRK).
@Oklahoma Schooners (Matt) :OK: acquire Jose Soriano from @McKinney Bay Knights (Ashtin) :MCK: in exchange for $2.00 of 2025 Griffey Coin.
7/17/2024 12:59 PM
@Andre - LA Smog :LA: acquire Lamonte Wade Jr. from @Fort Worth Fire (Austin) :FW: in exchange for $1.00 of 2025 Griffey Coin.
7/17/2024 2:07 PM
@Wachusett Mountaineers :COL: acquire Lane Thomas from @Princeton Anglers (GM - Dain) :PRI: in exchange for prospect Jett Williams and $5.00 of 2025 Griffey Coin.
7/17/2024 2:48 PM
@Sanger Vientos - Brandon :DAL: acquire Seiya Suzuki from @Princeton Anglers (GM - Dain) :PRI: in exchange for Ezequiel Tovar, Zach Neto, prospect Edgar Quero and $10.00 of 2025 Griffey Coin.
7/17/2024 2:53 PM
@Chicago Communists (Keith) :CHI: acquire Brandon Lowe from @Beamsville Beavers (Jacob) :SC: in exchange for $10.00 of 2025  Griffey Coin
7/17/2024 3:32 PM
@Chicago Communists (Keith) :CHI: acquire Dean Kremer from @Wachusett Mountaineers :COL:  in exchange for $1.00 of 2025 Griffey Coin
7/17/2024 3:40 PM
@Brooklyn Brewers (Nic) :BRK: acquire  Alec Burleson from @McKinney Bay Knights (Ashtin) :MCK: in exchange for a 2025 1st Rd OP (SC).
7/18/2024 7:38 AM
@Halton Huskies (Brendan) :HAL: acquire George Kirby, Bryce Miller and Walker Martin from @Amherst Minutemen :AMH: in exchange for Vinnie Pasquantino, Lazaro Montes and 2025 1st Round OP.
7/18/2024 8:22 PM
@Sanger Vientos - Brandon :DAL: acquire Jake Irvin from @Fort Worth Fire (Austin) :FW: in exchange for $4.00 of 2025 Griffey Coin. 
@Sanger Vientos - Brandon :DAL: acquire Spencer Horowitz, Jose Miranda (DFA), and a 2025 3rd Rd OP (MCK) from @McKinney Bay Knights (Ashtin) :MCK: in exchange for Austin Wynns, a 2025 1st Rd OP (DAL) and a 2025 2nd Rd OP (DAL).
7/18/2024 8:30 PM
@McKinney Bay Knights (Ashtin) :MCK: acquire Jake Burger (DFA) from @Wachusett Mountaineers :COL: in exchange for $1.00 of 2025 Griffey Coin.
@McKinney Bay Knights (Ashtin) :MCK: acquire Paul Goldschmidt (DFA) from @Brooklyn Brewers (Nic) :BRK: in exchange for $1.00 of 2025 Griffey Coin.
7/18/2024 8:37 PM
@McKinney Bay Knights (Ashtin) :MCK: acquire Cal QuantriWW (DFA) from @Beamsville Beavers (Jacob) :SC: in exchange for $1.00 of 2025 Griffey Coin and a 2025 4th Rd OP (MCK). 
7/18/2024 8:45 PM
@Sanger Vientos - Brandon :DAL: acquire Michael Lorenzen from @Princeton Anglers (GM - Dain) :PRI: in exchange for a 2025 3rd Rd OP (MCK) and a 2025 3rd Rd OP (DAL).
@McKinney Bay Knights (Ashtin) :MCK: acquire Michael "The Pete Getter" Soroka from @Princeton Anglers (GM - Dain) :PRI: in exchange for future considerations.
7/18/2024 9:23 PM
@Beamsville Beavers (Jacob) :SC: acquire Michael King from @Zach :NO: in exchange for prospect Alex Canario and $15.00 of 2025 Griffey Coin.
7/18/2024 9:56 PM
@Wachusett Mountaineers :COL: acquire Meatball Marco from @Amherst Minutemen :AMH: in exchange for $1 GC.
7/18/2024 11:11 PM
In a 3 team trade:
@Wachusett Mountaineers :COL: acquire Charlie Blackmon and prospect Jhonkeksy Noel from @Brooklyn Brewers (Nic) :BRK: 
@Brooklyn Brewers (Nic) :BRK: acquire Clayton Kershaw from @Beamsville Beavers (Jacob) :SC:
@Beamsville Beavers (Jacob) :SC: acquire Jordan Montgomery from @Wachusett Mountaineers :COL:. Montgomery has a $1.50 salary and expires in 2027
@Brooklyn Brewers (Nic) :BRK: acquire prospect George Lombard Jr. from @Amherst Minutemen :AMH: in exchange for $5 2026 GC.
1/11/2025 2:05 PM
@Amherst Minutemen :AMH: acquire Ketel Marte and prospect Walker Jenkins from @Andre - LA Smog :LA: in exchange for Junior Caminero.
1/11/2025 4:00 PM
@Toronto Canucks :TOR: acquire prospects James Triantos and Jacob Misiorowski from @Amherst Minutemen :AMH: in exchange for the rights to Bo Bichette.
1/11/2025 4:25 PM
@Wachusett Mountaineers :COL: acquire prospects Owen Caissie and Brock Wilken from @Amherst Minutemen :AMH: in exchange for the rights to Jarren Duran
1/14/2025 4:31 PM
@Brooklyn Brewers (Nic) :BRK: acquire prospect Rayner Arias from @Amherst Minutemen :AMH: in exchange for rights to Alex Bregman.
1/15/2025 3:27 PM
@Andre - LA Smog :LA: acquire Bryce Harper COL retains $5 from @Wachusett Mountaineers :COL: in exchange for Tanner Bibee, Gavin Williams, Michael Busch, Nate Lowe, Thairo Estrada, Yoan Moncada, Josh Rojas, prospect Xavier Isaac, and FA rights to Carlos Correa
1/15/2025 8:20 PM
@Amherst Minutemen :AMH: acquire the FA rights to Lawrence Butler from @Andre - LA Smog :LA: in exchange for $1 GC
1/16/2025 7:46 AM
@Wachusett Mountaineers :COL: acquire Curtis Mead from @Amherst Minutemen :AMH: in exchange for the FA rights to Jurisckson Profar.
1/16/2025 8:44 AM
@Amherst Minutemen :AMH: acquire the FA rights to Tanner Houck from @Andre - LA Smog :LA: in exchange for $1 GC
1/19/2025 5:24 PM
@Zach :NO: acquire Dustin May, 1st Rd OP (BRK), $11 2025 GC, $20 2026 GC, FA rights to Brandon Nimmo, Tyler Mahle and Kumar Rocker from @Andre - LA Smog :LA: in exchange for no return. 
1/21/2025 8:10 AM
@Beamsville Beavers (Jacob) :SC: acquire Gleyber Torres (LA retains $0.75) from @Andre - LA Smog :LA: in exchange for FA rights to Blake Snell
1/26/2025 6:57 AM
@Andre - LA Smog :LA: acquire $4 2025 GC from @Beamsville Beavers (Jacob) :SC: in exchange for $5 2026 GC
1/26/2025 4:56 PM
@Halton Huskies (Brendan) :HAL: acquire $20 2025 GC from @Beamsville Beavers (Jacob) :SC: in exchange for $25 2026 GC
1/28/2025 11:53 AM
@McKinney Bay Knights (Ashtin) :MCK: acquire Jazz Jizzholm Jr and Free Agent rights to Austin Riley from @Fort Worth Fire (Austin) :FW: in exchange for Andy Pages, a 2nd Rd OP (BRK - 25th Overall), and $163 of 2025 Griffey Coin.
1/28/2025 11:54 AM
@Beamsville Beavers (Jacob) :SC: acquire Brenton Doyle rights from @Andre - LA Smog :LA: in exchange for $1 of 2025 Griffey Coin. 
@Fort Worth Fire (Austin) :FW: acquire the rights to Brendan Rodgers, Keibert Ruiz, Ryan Mountcastle, Victor Robles, Jacob Stallings, DJ Herz, Landon Knack, Brant Hurter, and Vaughn Grissom from @Andre - LA Smog :LA:  for $1 of 2025 Griffey Coin. 
1/28/2025 1:28 PM
@Princeton Anglers (GM - Dain) :PRI: acquire alleged pedophile Wander Franco, Henry Davis, and prospect Zac Veen from @McKinney Bay Knights (Ashtin) :MCK: in exchange for not a god damn thing.
1/28/2025 2:12 PM
@Beamsville Beavers (Jacob) :SC: acquire Garret Mitchell from @Chicago Communists (Keith) :CHI: in exchange for $10 2025 GC.
1/29/2025 12:39 PM
@Amherst Minutemen :AMH: acquire Jacob DeGrom and $80 2025 GC from @Sanger Vientos - Brandon :SGR: in exchange for Bretty Baty, prospects Harry Ford, Christian Scott, Samuel Zavala, OP 1-12 (12th overall) and the FA rights to Spencer Torkelson
Condition: :AMH: sends :SGR: $10 2026 GC if Jacob DeGrom makes at least 20 starts. 
1/29/2025 1:40 PM
@Oklahoma Schooners (Matt) :OK: acquire Jeimer Candelario and $3 2025 GC from @Amherst Minutemen :AMH: in exchange for no return.
1/29/2025 5:54 PM
@Amherst Minutemen :AMH: acquire Manny Machado from @Beamsville Beavers (Jacob) :BEA: in exchange for Edouard Julien, Emerson Hancock, the FA rights to Alex Bregman, and $5 2025 GC.
1/30/2025 8:24 AM
@Beamsville Beavers (Jacob) :BEA: acquire prospects Thomas Saggesse and Brock Porter from @Fort Worth Fire (Austin) :FW: in exchange for $10 GC
1/30/2025 10:54 AM
@Beamsville Beavers (Jacob) :BEA: acquire prospect Gavin Cross from @Fort Worth Fire (Austin) :FW: in exchange for $3 GC
1/30/2025 2:13 PM
@Amherst Minutemen :AMH: acquire Luis Castillo from @Beamsville Beavers (Jacob) :BEA: in exchange for Kutter Crawford and the FA rights to Brayan Bello and Tanner Houck
1/31/2025 3:19 PM
@Toronto Canucks :TOR: acquire prospect Addison Barger from @Beamsville Beavers (Jacob) :BEA: in exchange for $1 2025 GC
2/11/2025 6:49 PM
@Andre - LA Smog :LA: acquire $7 of 2026 GC from @Zach :NO: in exchange for $5 of 2025 GC.
2/12/2025 7:58 AM
@Beamsville Beavers (Jacob) :BEA: acquire FA rights to Nick Lodolo and $2 2026 GC from @Zach :NO: in exchange for $10 2025 GC.
2/13/2025 1:38 PM
@Zach :NO: acquire $23 GC 2025 from @Beamsville Beavers (Jacob) :BEA: in exchange for $23 GC 2026
2/16/2025 8:20 AM
@Amherst Minutemen :AMH: acquire the FA rights to Logan Webb and Robbie Ray from @Brooklyn Brewers (Nic) :BRK: in exchange for $15 GC.
3/6/2025 10:04 AM
@Andre - LA Smog :LA: acquire prospect Zyhir Hope and Joe Musgrove from @Toronto Canucks :TOR: in exchange for no return.
3/20/2025 8:27 AM
@Oklahoma Schooners (Matt) :OK: acquire $1 GC and the FA rights to Wilyer Abreu, Michael Massey, Mitchell Parker, Jonathan Cannon, Keider Montero, Davis Martis, Jake Bloss, Ky Bush from @Amherst Minutemen :AMH: in exchange for no return. 
3/20/2025 11:42 AM
@Amherst Minutemen :AMH: acquire prospect Denzel Clarke and OP 3-14 from @Andre - LA Smog :LA: in exchange for OP 3-3.
3/26/2025 11:14 AM
@Amherst Minutemen :AMH: acquire Andy Pages and $10 GC from @Fort Worth Fire (Austin) :FW: in exchange for Willy Adames 
3/26/2025 1:23 PM
@Oklahoma Schooners (Matt) :OK: acquire Matt Chapman from @McKinney Bay Knights (Ashtin) :MCK: in exchange for Kyle Manzardo.
3/30/2025 11:34 PM
@Zach :NO: acquire Tony Gonsolin from @Sanger Vientos - Brandon :SGR: in exchange for prospect Marco Luciano.
4/1/2025 8:47 AM
@Halton Huskies (Brendan) :HAL: acquire Luis Castillo from @Amherst Minutemen :AMH: in exchange for Justin Verlander and $10 GC.
4/1/2025 9:35 AM
@Amherst Minutemen :AMH: acquire Jordan Hicks from @Fort Worth Fire (Austin) :FW: in exchange for Mitchell Parker, prospect Denzel Clarke and $3 GC. 
4/1/2025 9:02 PM
@Halton Huskies (Brendan) :HAL: acquire Emerson Hancock from @Beamsville Beavers (Jacob) :BEA: in exchange for $3 GC 
4/3/2025 12:03 PM
@McKinney Bay Knights (Ashtin) :MCK: acquire Tarik Skubal at 8.50 retained from @Wachusett Mountaineers :COL: in exchange for Jordan Walker, and for prospects Jasson Dominguez, Carson Williams, and Ethan Salas. 
4/3/2025 9:08 PM
@Amherst Minutemen :AMH: acquire Salvador Perez from @Sanger Vientos - Brandon :SGR: in exchange for Andy Pages.
4/4/2025 9:25 AM
@Beamsville Beavers (Jacob) :BEA: acquire Ben Rice and a 2026 Smog 4th round open draft pick from @Andre - LA Smog :LA: in exchange for $7 GC and a 2026  Beaver 3rd round open draft pick.
4/6/2025 7:56 AM
@Andre - LA Smog :LA: acquire Gabriel Arias from @Wachusett Mountaineers :COL: in exchange for $1 GC.
4/7/2025 9:06 AM
@Oklahoma Schooners (Matt) :OK: acquire Mike Yastremski from @McKinney Bay Knights (Ashtin) :MCK:  in exchange for $1 GC. 
4/7/2025 11:04 AM
@Andre - LA Smog :LA: acquire Matthew Liberatore from @Zach :NO: in exchange for $1 GC.
4/8/2025 3:00 PM
@Fort Worth Fire (Austin) :FW: acquire Pete Alonso from @Sanger Vientos - Brandon :SAN: in exchange for Jake Burger and prospects JJ Wetherholt, Kevin Alcantara and Cam Collier. Sanger retains $3.25 per year on Alonso.
4/10/2025 5:00 AM
@Amherst Minutemen :AMH: acquire Lars Nootbaar, Sonny Gray and Shea Langeliers from @Wachusett Mountaineers :COL: in exchange for Hunter Goodman and prospects Chase DeLauter and Tink Hence
4/14/2025 9:34 AM
@Wachusett Mountaineers :WAC: acquire Gabriel Arias from @Andre - LA Smog :LA: in exchange for $1 GC.
4/16/2025 10:46 AM
@Sanger Vientos - Brandon :SAN: acquire Bobby Miller from @Andre - LA Smog :LA: in exchange for $1 GC.
4/17/2025 7:49 AM
@Fort Worth Fire (Austin) :FW: acquire J.P. Sears from @Andre - LA Smog :LA: in exchange for Easton Lucas and $1 GC.
4/18/2025 12:01 PM
@Beamsville Beavers (Jacob) :BEA: acquire Eduardo Rodriguez and a 2026 2nd RD Open Pick (WAC) from @Wachusett Mountaineers :WAC: in exchange for $10 2026 GC and a 2026 2nd RD Open Pick (BEA)
4/25/2025 12:58 PM
@Andre - LA Smog :LA: acquire Wilyer Abreu and Tyler Sodastream, as well as $3 GC, from the @Sanger Vientos - Brandon :SAN:  in exchange for Evan Carter, and prospects Zyhir Hope, Franklin Arias and Aiden Smith. 
4/26/2025 11:26 PM
@Andre - LA Smog :LA: acquire Noelvi Marte from @Fort Worth Fire (Austin) :FW:  in exchange for $6 GC.
4/27/2025 2:50 PM
@McKinney Bay Knights (Ashtin) :MCK: acquire Austin Hays (DFA) from @Princeton Anglers (GM - Dain) :PRI: in exchange for Justyn-Henry Malloy and prospect Heston Kjerstad.
4/29/2025 6:06 PM
@Amherst Minutemen :AMH: acquire Jorge Polanco from @Sanger Vientos - Brandon :SAN: in exchange for $5 GC and Kyle Stowers
@Brooklyn Brewers (Nic) :BRK: acquire Clarke Schmidt from @Zach :NO: in exchange for $4 GC.
5/5/2025 8:22 AM
@Beamsville Beavers (Jacob) :BEA: acquire Luis Arraez from @Wachusett Mountaineers :WAC: in exchange for Bryson Stott and prospects Carson Benge, Enrique Bradfield Jr, and Nick Yorke
5/5/2025 9:19 AM
@Brooklyn Brewers (Nic) :BRK: acquire  Gunner Hoglund from @Wachusett Mountaineers :WAC: in exchange for $5 2026 GC
5/5/2025 9:48 AM
@Fort Worth Fire (Austin) :FW: acquire Bryson Stott from @Wachusett Mountaineers :WAC: in exchange for Zack Gelof and $20 GC 
5/6/2025 8:30 AM
@Halton Huskies (Brendan) :HAL: acquire Anthony Santander from @Wachusett Mountaineers :WAC: in exchange for prospect Michael Arroyo
5/6/2025 11:12 AM
@Andre - LA Smog :LA: acquire Shea Langeliers from @Amherst Minutemen :AMH: in exchange for $1 GC.
5/7/2025 10:21 AM
@Sanger Vientos - Brandon :SAN: acquire Jhonkensy Noel, Curtis Mead, Carlos Correa, prospect Jefferson Rojas, and $8 GC from @Wachusett Mountaineers :WAC: in exchange for not having those guys on Patrick's team anymore.
5/8/2025 3:53 PM
@Toronto Canucks :TOR: acquire Garrett Crochet and Shea Langeliers from @Andre - LA Smog :LA: in exchange for prospects Bryce Eldridge, Arjun Nimmala, Jacob Misiorowski, Ralphy Velasquez, and $10 GC.
5/8/2025 6:49 PM
@Andre - LA Smog :LA: acquire Tanner Bibee and Gavin Williams from @Wachusett Mountaineers :WAC: in exchange for prospects Max Muncy (ATH) and Brice Matthews. 
5/9/2025 9:19 PM
@Amherst Minutemen :AMH: acquire Matthew Liberatore from @Andre - LA Smog :LA: in exchange for Jorge Polanco.
5/12/2025 8:11 AM
@McKinney Bay Knights (Ashtin) :MCK: acquire Cal Quantrill from @Beamsville Beavers (Jacob) :BEA: in exchange for $1 GC.
5/19/2025 8:43 AM
@Amherst Minutemen :AMH: acquire Javier Baez from @Wachusett Mountaineers :WAC: in exchange for $2 of 2026 GC.
5/19/2025 9:12 PM
@McKinney Bay Knights (Ashtin) :MCK: acquire Francisco Lindor and Jose Altuve from @Chicago Communists (Keith) :CHI: in exchange for prospect Marcelo Mayer
5/19/2025 9:37 PM
@Oklahoma Schooners (Matt) :OK: acquire Colton Gordon from @McKinney Bay Knights (Ashtin) :MCK: in exchange for $2 GC.
5/20/2025 7:42 AM
@Amherst Minutemen :AMH: acquire Corbin Burnes, Marcel Ozuna, Eric Fedde, and Willson Contreras from @Chicago Communists (Keith) :CHI: in exchange for Vinnie Pasquintino, Alec Burleson, Landon Ruopp, prospect Walker Jenkins, and $8 of 2026 GC 
5/20/2025 3:59 PM
@Andre - LA Smog :LA: acquire Hunter Goodman from @Wachusett Mountaineers :WAC: in exchange for prospect Colby Thomas.
5/20/2025 6:45 PM
@Andre - LA Smog :LA: acquire Matthew Boyd from @Chicago Communists (Keith) :CHI: in exchange for $10 GC.
5/21/2025 5:29 PM
@Beamsville Beavers (Jacob) :BEA: acquire Max Muncy from @Chicago Communists (Keith) :CHI: in exchange for Miguel Vargas.
5/22/2025 3:37 PM
@Wachusett Mountaineers :WAC: acquire Reynaldo Lopez and prospects Jac Caglianone, Sam Basallo, Charlie Condon, Denzel Clarke from @Fort Worth Fire (Austin) :FW: in exchange for Paul Skenes, Mitch Keller and Hunter Dobbins
5/23/2025 10:49 AM
@Beamsville Beavers (Jacob) :BEA: acquire Yusei Kikuchi from @Halton Huskies (Brendan) :HAL: in exchange for $5 of 2026 GC
5/27/2025 7:59 AM
@McKinney Bay Knights (Ashtin) :MCK: acquire Yoan Moncada from @Wachusett Mountaineers :WAC: in exchange for $1 of GC.
5/27/2025 10:40 AM
@Wachusett Mountaineers :WAC: acquire Brett Baty from @Sanger Vientos - Brandon :SAN: in exchange for $5 of 2026 GC.
5/29/2025 12:01 PM
@Wachusett Mountaineers :WAC: acquire Miguel Vargas from @Chicago Communists (Keith) :CHI: in exchange for $20 of 2026 GC.
5/29/2025 9:32 PM
@Amherst Minutemen :AMH: acquire Julio Rodriguez (MCK retains $2) from @McKinney Bay Knights (Ashtin) :MCK: in exchange for Lars Nootbar and Jacob deGrom
5/30/2025 9:32 AM
@McKinney Bay Knights (Ashtin) :MCK: acquire Issac Paredes, Carson Kelly, and $10 GC from @Sanger Vientos - Brandon :SAN: in exchange for Ryan Jeffers, Yoan Moncada, Trevor Story, Colt Keith, prospect Orelvis Martinez, prospect Druw Jones, and prospect Ricky Tiedemann.
5/30/2025 12:31 PM
@Amherst Minutemen :AMH: acquire Jameson Taillon from @Chicago Communists (Keith) :CHI: in exchange for $4 GC.
5/31/2025 8:22 AM
@Zach :NO: acquire Patrick Corbin from @Chicago Communists (Keith) :CHI: in exchange for $2 of 2026 GC.
5/31/2025 1:35 PM
@Amherst Minutemen :AMH: acquire prospect Harry Ford from @Sanger Vientos - Brandon :SAN: in exchange for prospect Joendry Vargas and $5 GC 
6/2/2025 5:23 AM
@Amherst Minutemen :AMH: acquire Tylor Megill from @Oklahoma Schooners (Matt) :OK: in exchange for prospect Jaison Chourio.
6/2/2025 6:43 PM
@Andre - LA Smog :LA: acquire MacKenzie Gore, Edward Cabrera and prospect Brailer Guerrero from @Chicago Communists (Keith) :CHI: in exchange for prospects Bryce Eldridge, Emmanuel Rodriguez, and Jacob Misiorowski
6/3/2025 12:33 PM
@Andre - LA Smog :LA: acquire Lance McCullers Jr. from @Wachusett Mountaineers :WAC: in exchange for $2 GC.
6/4/2025 9:37 PM
@Zach :NO: acquire Landon Knack from @Chicago Communists (Keith) :CHI: in exchange for $1 of 2026 GC.
6/5/2025 10:46 AM
@Sanger Vientos - Brandon :SAN: acquire Clayton Kershaw and a 6th round Open Draft pick from @Andre - LA Smog :LA: in exchange for obtaining a 5th round open draft pick. 
6/9/2025 1:33 PM
@Beamsville Beavers (Jacob) :BEA: acquire Tyler O'Neill from @Fort Worth Fire (Austin) :FW: in exchange for $6 of 2026 Griffey coin
6/12/2025 9:29 AM
@Toronto Canucks :TOR: acquire Tomoyuki Sugano from @Amherst Minutemen :AMH: in exchange for $3 of 2026 Griffey coin.
6/13/2025 7:11 PM
@Brooklyn Brewers (Nic) :BRK: acquire Stephen Kolek from @Chicago Communists (Keith) :CHI: in exchange for $2 of 2026 GC.
6/15/2025 10:10 AM
@Andre - LA Smog :LA: acquire prospects Cam Collier, Zyhir Hope, and Franklin Arias from @Sanger Vientos - Brandon :SAN: in exchange for $36 of 2026 GC.
6/16/2025 9:02 AM
@Andre - LA Smog :LA: acquire Emmet Sheehan from @Fort Worth Fire (Austin) :FW: in exchange for Matthew Boyd. @Fort Worth Fire (Austin) retains 1.50 on Sheehan's salary.
6/16/2025 12:21 PM
@Amherst Minutemen :AMH: acquire Quinn Priester from @Wachusett Mountaineers :WAC: in exchange for $1 GC.
6/16/2025 1:34 PM
@Fort Worth Fire (Austin) :FW: acquire Jeff McNeil from @Halton Huskies (Brendan) :HAL: in exchange for $1 GC.
6/24/2025 8:21 PM
@Toronto Canucks :TOR: acquire Nathan Eovaldi, Royce Lewis, prospect Kevin Alcantara and prospect Druw Jones from @Sanger Vientos - Brandon :SAN:  in exchange for Josh Jung, Gabriel Moreno, prospect Luis Angel Acuna and prospect Robert Calaz. @Sanger Vientos - Brandon retains $1.50 on Eovaldi's salary
6/26/2025 3:31 PM
@Andre - LA Smog :LA: acquire Clayton Kershaw from @Sanger Vientos - Brandon :SAN:  for prospect Ralphy Velazquez. 
6/27/2025 8:51 PM
@Amherst Minutemen :AMH: acquire Juan Soto from @Brooklyn Brewers (Nic) :BRK: in exchange for Jackson Merrill and prospects Colt Emerson and Jesus Made
6/29/2025 10:03 PM
@Princeton Anglers (GM - Dain) :PRI: acquire Jose Ramirez, Ozzie Albies, Carlos Rodon, and Clarke Schmidt from @Brooklyn Brewers (Nic) :BRK:  in exchange for Jackson Chourio and prospects Nick Kurtz, Jordan Lawler, and Andrew Painter. 
6/30/2025 9:42 AM
@Princeton Anglers (GM - Dain) :PRI: acquire Adrian Houser and a 2026 3rd round OP (CHI) from @Chicago Communists (Keith) :CHI: in exchange for a 2026 2nd round OP (PRI)
@Zach :NO: acquire Lucas Giolito from @Chicago Communists (Keith) :CHI: in exchange for $2 of 2026 GC.
6/30/2025 9:57 AM
@Toronto Canucks :TOR: acquire Alec Burleson from @Chicago Communists (Keith) :CHI: in exchange for $2 of 2026 GC.
6/30/2025 1:26 PM
@Amherst Minutemen :AMH: acquire Landon Ruopp from @Chicago Communists (Keith) :CHI: in exchange for $1 of 2026 GC.
7/6/2025 10:58 AM
@Wachusett Mountaineers :WAC: acquire Joe Boyle from @Andre - LA Smog :LA: in exchange for $3 of 2026 GC.
7/6/2025 1:38 PM
@Amherst Minutemen :AMH: acquire Trevor Story from @Sanger Vientos - Brandon :SAN: in exchange for Bo Bichette and $2 GC 
7/6/2025 9:15 PM
@Princeton Anglers (GM - Dain) :PRI: acquire Randy Arozarena from @Chicago Communists (Keith) :CHI: in exchange for prospect Jett Williams.
7/7/2025 9:38 AM
@Halton Huskies (Brendan) :HAL: acquire Logan Gilbert and Juan Brito from @Chicago Communists (Keith) :CHI: in exchange for Gerrit Cole and prospects Travis Sykora and Josue Briceno and $15 of 2025 Griffey Coin. 
7/7/2025 11:26 AM
@Beamsville Beavers (Jacob) :BEA: acquire Cedanne Rafaela from @Chicago Communists (Keith) :CHI: in exchange for $15 of 2025 griffey coin.
7/8/2025 12:24 PM
@Amherst Minutemen :AMH: acquire Gavin Williams from @Andre - LA Smog :LA: in exchange for $1 GC
7/9/2025 9:06 AM
@Fort Worth Fire (Austin) :FW: acquire Matt Chapman from @Oklahoma Schooners (Matt) :OK: in exchange for prospect Jackson Jobe.
7/9/2025 10:22 AM
@Fort Worth Fire (Austin) :FW: acquire Nico Hoerner, Ryne Nelson (DFA), and $10.00 of GC from @Brooklyn Brewers (Nic) :BRK: in exchange for Jackson Holiday and Austin Riley. Fort Worth retains $3.00 of Riley's annual salary.
7/9/2025 2:59 PM
@Brooklyn Brewers (Nic) :BRK: acquire Cam Schlitter and $5 of GC from @Wachusett Mountaineers :WAC: in exchange for prospects Kaelen Culpepper and Braylon Payne.
7/10/2025 10:59 AM
@Zach :NO: acquire Bryan Bello, Hunter Dobbins, and Walker Buehler from @Fort Worth Fire (Austin) :FW: in exchange for Roki Sasaki, Patrick Corbin, and prospect Matt Shaw. New Orleans retains $5.00 of Sasaki's annual salary.
7/10/2025 11:37 AM
@Zach :NO: acquire Trevor Rogers from @Wachusett Mountaineers :WAC: in exchange for prospect Roderick Arias.
7/10/2025 2:53 PM
@McKinney Bay Knights (Ashtin) :MCK: acquire Aaron Judge from @Brooklyn Brewers (Nic) :BRK: in exchange for Pete Crow-Armstrong, Brandon Lowe, Luis Severino, prospects Ronny Mauricio and Spencer Jones; $35 of GC and a 2026 1st Rd OP (MCK).
7/10/2025 5:59 PM
@Andre - LA Smog :LA: acquire Luis Garcia Jr. from @McKinney Bay Knights (Ashtin) :MCK: in exchange for prospect Ryan Waldschmidt.
7/10/2025 9:14 PM
@Fort Worth Fire (Austin) :FW: acquire Ernie Clement from @Chicago Communists (Keith) :CHI: in exchange for Jake Cronenworth and $1 of 2026 GC.
7/11/2025 1:13 PM
@Zach :NO: acquire Trea Turner, Luis Arraez, and Casey Mize from @Beamsville Beavers (Jacob) :BEA: in exchange for Dustin May, Brice Turang, and prospects  Travis Bazzana, Bryce Rainer, and Hagen Smith
7/11/2025 3:45 PM
@Andre - LA Smog :LA: acquire Gleyber Torres from @Beamsville Beavers (Jacob) :BEA: in exchange for Tanner Bibee, Riley Greene, and prospect CJ Kayfus.
7/13/2025 12:07 AM
@McKinney Bay Knights (Ashtin) :MCK: acquire Chris Sale from @Brooklyn Brewers (Nic) :BRK: in exchange for prospect Ryan Waldschmidt and $10 GC.
7/14/2025 10:41 AM
@Beamsville Beavers (Jacob) :BEA: acquire Marcel Ozuna from @Amherst Minutemen :AMH: in exchange for $1 GC.
7/14/2025 1:57 PM
@Beamsville Beavers (Jacob) :BEA: acquire CJ Abrams, Royce Lewis, prospect Kevin Alcantara and prospect Gabriel Gonzalez from @Toronto Canucks :TOR: in exchange for Freddie Freeman, Seth Lugo and Geraldo Perdomo
7/14/2025 2:12 PM
@Beamsville Beavers (Jacob) :BEA: acquire Nick Martinez from @Chicago Communists (Keith) :CHI: in exchange for $2 GC.
7/14/2025 5:51 PM
@Amherst Minutemen :AMH: acquire Romy Gonzalez from @Sanger Vientos - Brandon :SAN: in exchange for $3 GC.
7/15/2025 10:24 AM
@Beamsville Beavers (Jacob) :BEA: acquire Ryan O Hearn from @Toronto Canucks :TOR: in exchange for $3 GC
7/15/2025 11:24 AM
@Zach :NO: acquire $1 GC from @Wachusett Mountaineers :WAC:  in exchange for prospect Won Bin Cho 
7/15/2025 12:12 PM
@Oklahoma Schooners (Matt) :OK:  acquire Lance McCullers Jr. from @Andre - LA Smog :LA:  in exchange for $1 GC.
7/15/2025 12:36 PM
@Amherst Minutemen :AMH: acquire Frankie Montas from @Oklahoma Schooners (Matt) :OK: in exchange for $1 GC.
7/15/2025 5:37 PM
@Amherst Minutemen :AMH: acquire Kodai Senga and Matthew Boyd from @Fort Worth Fire (Austin) :FW: in exchange for Kristian Campbell, Jared Jones, and Quinn Priester. 
Trade Condition: If Kodai Senga does not go on the 15/60 day IL for the remainder of 2025, AMH will send FW $10 GC.
7/15/2025 5:39 PM
@Brooklyn Brewers (Nic) :BRK: acquire Ree Olsen from @Sanger Vientos - Brandon :SAN: in exchange for Jurickson PED Profar and Spencer Steer.
7/16/2025 1:17 PM
@Sanger Vientos - Brandon :SAN: acquire Michael Busch and prospects Jasson Dominguez, Carson Williams Chase Delauter, Brice Matthews, and Colby Thomas from @Wachusett Mountaineers :WAC: in exchange for prospects JJ Wetherholt, Aidan Smith, Joendry Vargas, and Jefferson Rojas.
7/16/2025 3:31 PM
@McKinney Bay Knights (Ashtin) :MCK: acquire Quinn Priester from @Fort Worth Fire (Austin) :FW: in exchange for $8 GC.
7/16/2025 7:36 PM
@Beamsville Beavers (Jacob) :BEA: acquire Zac Gallen, Zach Eflin, Taylor Ward, and $10 GC from @Brooklyn Brewers :BRK: in exchange for Ben Rice, Dustin May and prospect Sal Stewart
7/17/2025 6:08 AM
@Fort Worth Fire (Austin) :FW: acquire Marc Liberatore from @Amherst Minutemen :AMH: in exchange for $2 GC
7/17/2025 9:47 AM
@Andre - LA Smog :LA: acquire Brent Rooker from @Sanger Vientos - Brandon :SAN:  in exchange for Noelvi Marte, Luis Garcia Jr., prospect Cam Collier, and the Smog 2026 second round open draft pick. 
7/17/2025 9:52 AM
@Beamsville Beavers (Jacob) :BEA: acquire Dean Kremer, prospect Carson Whisenhunt and $2 GC from @Fort Worth Fire (Austin) :FW: in exchange for Josh Smith and Brady Singer (from DFA team)
7/17/2025 9:56 AM
@Chicago Communists (Keith) :CHI: acquire Seiya Suzuki and Colt Keith from @Sanger Vientos - Brandon :SAN: in exchange for Jung-Hoo Lee,  prospects Jett Williams and Emmanuel Rodriguez, and $15 of GC.
7/17/2025 10:03 AM
@Halton Huskies (Brendan) :HAL: acquire Bo Bichette, Josh Jung and a 2026 2nd rd OP (SAN) from @Sanger Vientos - Brandon :SAN: in exchange for prospect Dalton Rushing, prospect Alex Freeland, a 2026 1st rd OP (HAL) and $3 GC. 
7/17/2025 1:06 PM
@Amherst Minutemen :AMH: acquire Jonathan Cannon from @Chicago Communists (Keith) :CHI: in exchange for $1 GC.
7/17/2025 2:12 PM
@Toronto Canucks :TOR: acquire Joe Ryan from @Sanger Vientos - Brandon :SAN: in exchange for Addison Barger, David Festa, $35 GC and a 2026 1st Rd OP (TOR)
7/17/2025 2:35 PM
@Halton Huskies (Brendan) :HAL: acquire Anthony Volpe and $2 GC from @Fort Worth Fire (Austin) :FW: in exchange for Quinn Mathews

"""

def parse_trades(text):
    # Split lines but keep them in a list
    lines = text.strip().split('\n')
    rows = []
    
    date_pattern = re.compile(r'^\d{1,2}/\d{1,2}/\d{4}')
    team_pattern = re.compile(r'@(.*?)\s*:(\w+):')

    current_date = None
    current_time = None

    # UPDATED: enumerate(lines, start=1) gives us a counter 'i' starting at 1
    for i, line in enumerate(lines, start=1):
        line = line.strip()
        if not line: continue

        if date_pattern.match(line):
            try:
                parts = line.split(' ', 1)
                current_date = parts[0]
                current_time = parts[1] if len(parts) > 1 else ""
            except:
                pass 
            continue
        
        if line.startswith('@'):
            try:
                # 1. Separate Acquirer
                parts_1 = line.split(' acquire ', 1)
                acquirer_str = parts_1[0]
                rest = parts_1[1]
                
                # 2. Separate Sender
                parts_2 = rest.split(' from ', 1)
                assets_acquired_str = parts_2[0]
                rest_2 = parts_2[1]
                
                # 3. Separate Return
                if ' in exchange for ' in rest_2:
                    parts_3 = rest_2.split(' in exchange for ', 1)
                    sender_str = parts_3[0]
                    assets_returned_str = parts_3[1]
                elif '(no return)' in rest_2:
                    sender_str = rest_2.replace('(no return)', '').replace('.', '').strip()
                    assets_returned_str = None
                else:
                    sender_str = rest_2
                    assets_returned_str = None

                acquirer_match = team_pattern.search(acquirer_str)
                sender_match = team_pattern.search(sender_str)
                
                if acquirer_match and sender_match:
                    acquirer_id = acquirer_match.group(2)
                    sender_id = sender_match.group(2)    
                else:
                    # If we can't find a team ID, raise an error to trigger the exception block
                    raise ValueError("Could not find Team ID in: " + line[:20])

                def clean_assets(asset_string):
                    if not asset_string: return []
                    s = asset_string.rstrip('.')
                    s = re.sub(r', and ', '|', s)
                    s = re.sub(r' and (?!.*(?:Retain|retained))', '|', s)
                    s = s.replace(',', '|')
                    return [x.strip() for x in s.split('|') if x.strip()]

                acquired_items = clean_assets(assets_acquired_str)
                returned_items = clean_assets(assets_returned_str)

                for item in acquired_items:
                    rows.append({
                        'Date': current_date,
                        'Time': current_time,
                        'Team From': sender_id,
                        'Team To': acquirer_id,
                        'Item Traded': item
                    })

                if returned_items:
                    for item in returned_items:
                        rows.append({
                            'Date': current_date,
                            'Time': current_time,
                            'Team From': acquirer_id,
                            'Team To': sender_id,
                            'Item Traded': item
                        })

            # UPDATED EXCEPTION BLOCK
            except Exception as e:
                # 'i' is the line number from enumerate()
                print(f" Error on Line {i}: {e}")
                print(f"   Text: {line[:50]}...")
                continue

    return pd.DataFrame(rows)

df = parse_trades(raw_text)
print(df.to_csv(index=False))
```

```python
import sqlite3
import csv
import os

def export_sqlite_to_csv(database_path = "fantasy_app/pages/fantasy.db", output_directory="C:/Users/mattm/database/migration/"):
    """
    Exports all tables from an SQLite database to individual CSV files.

    Args:
        database_path (str): The path to the SQLite database file.
        output_directory (str): The directory where CSV files will be saved.
                                 Defaults to "exported_csv".
    """
    if not os.path.exists(output_directory):
        os.makedirs(output_directory)

    conn = None
    try:
        conn = sqlite3.connect(database_path)
        cursor = conn.cursor()

        # Get a list of all table names in the database
        cursor.execute("SELECT name FROM sqlite_master WHERE type='table' AND name ='game_stats';")
        tables = cursor.fetchall()

        for table_name_tuple in tables:
            table_name = table_name_tuple[0]
            csv_file_path = os.path.join(output_directory, f"{table_name}.csv")

            print(f"Exporting table '{table_name}' to '{csv_file_path}'...")

            # Select all data from the current table
            cursor.execute(f"SELECT * FROM {table_name}")
            rows = cursor.fetchall()

            # Get column headers
            headers = [description[0] for description in cursor.description]

            with open(csv_file_path, 'w', newline='', encoding='utf-8') as csvfile:
                csv_writer = csv.writer(csvfile)
                csv_writer.writerow(headers)  # Write headers
                csv_writer.writerows(rows)    # Write data rows

        print("Export complete.")

    except sqlite3.Error as e:
        print(f"SQLite error: {e}")
    except IOError as e:
        print(f"File I/O error: {e}")
    finally:
        if conn:
            conn.close()

# Example usage:
if __name__ == "__main__":
    db_file = "fantasy.db"  # Replace with your database file name
    export_sqlite_to_csv(db_file)
```

```python
import os
import shutil

source_dir = r'C:\Users\mattm\FSDB\database\player_photos'
files_per_folder = 100
# Optional: specify image extensions if you only want to move images
# image_extensions = ('.png', '.jpg', '.jpeg', '.gif', '.bmp') 

def batch_files(src_folder):
    files = [f for f in os.listdir(src_folder) if os.path.isfile(os.path.join(src_folder, f))]
    # Optional: filter by image extensions if needed
    # files = [f for f in files if f.lower().endswith(image_extensions)]

    i = 0
    folder_counter = 1
    curr_subdir = None

    for f in files:
        if i % files_per_folder == 0:
            subdir_name = os.path.join(src_folder, f'Batch_Folder_{folder_counter:03d}')
            os.makedirs(subdir_name, exist_ok=True) # Create new directory
            curr_subdir = subdir_name
            folder_counter += 1
        
        # Move file to current directory
        f_base = os.path.basename(f)
        shutil.move(os.path.join(src_folder, f), os.path.join(curr_subdir, f_base))
        i += 1
    
    print(f"Moved {i} files into {folder_counter - 1} folders.")

if __name__ == "__main__":
    batch_files(source_dir)
```
